```python
# ===================================================================
# –ò–¢–û–ì–û–í–ê–Ø –†–ê–ë–û–¢–ê: A/B –¢–ï–°–¢–ò–†–û–í–ê–ù–ò–ï –†–ê–ó–ú–ï–©–ï–ù–ò–Ø –†–ï–ö–õ–ê–ú–ù–û–ì–û –ë–õ–û–ö–ê
# –ö–£–†–°: "–ê–ù–ê–õ–ò–¢–ò–ö –î–ê–ù–ù–´–• –° –ù–£–õ–Ø"
# –í–ê–ñ–ù–û: CR = adds/clicks - ratio-–º–µ—Ç—Ä–∏–∫–∞, –∏—Å–ø–æ–ª—å–∑—É–µ–º –¥–µ–ª—å—Ç–∞-–º–µ—Ç–æ–¥
# ===================================================================

import os
import pandas as pd
import numpy as np
from scipy import stats
from scipy.stats import ttest_ind
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# –ù–∞—Å—Ç—Ä–æ–π–∫–∞ —Å—Ç–∏–ª—è
plt.style.use('seaborn-v0_8-whitegrid')
plt.rcParams['figure.figsize'] = (14, 10)
plt.rcParams['font.size'] = 11

print("="*80)
print("üéì –ò–¢–û–ì–û–í–ê–Ø –†–ê–ë–û–¢–ê: A/B –¢–ï–°–¢–ò–†–û–í–ê–ù–ò–ï –†–ê–ó–ú–ï–©–ï–ù–ò–Ø –†–ï–ö–õ–ê–ú–´")
print("="*80)
print("üìö –ö—É—Ä—Å: '–ê–Ω–∞–ª–∏—Ç–∏–∫ –¥–∞–Ω–Ω—ã—Ö —Å –Ω—É–ª—è'")
print("üéØ –¢–µ–º–∞: –≠–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç –ø–æ –∏–∑–º–µ–Ω–µ–Ω–∏—é –ø–æ–∑–∏—Ü–∏–∏ —Ä–µ–∫–ª–∞–º–Ω–æ–≥–æ –±–ª–æ–∫–∞")
print("‚ö†Ô∏è  –í–ê–ñ–ù–û: CR = adds/clicks - ratio-–º–µ—Ç—Ä–∏–∫–∞, –∏—Å–ø–æ–ª—å–∑—É–µ–º –¥–µ–ª—å—Ç–∞-–º–µ—Ç–æ–¥")
print("="*80)

# ===================================================================
# –ß–ê–°–¢–¨ 1: –ü–û–î–ì–û–¢–û–í–ö–ê –î–ê–ù–ù–´–• –ò –î–ò–ó–ê–ô–ù –≠–ö–°–ü–ï–†–ò–ú–ï–ù–¢–ê
# ===================================================================

print("\n" + "="*80)
print("üìä –ß–ê–°–¢–¨ 1: –ü–û–î–ì–û–¢–û–í–ö–ê –î–ê–ù–ù–´–• –ò –î–ò–ó–ê–ô–ù –≠–ö–°–ü–ï–†–ò–ú–ï–ù–¢–ê")
print("="*80)

# -------------------------------------------------
# 1.1 –ó–∞–≥—Ä—É–∑–∫–∞ –∏ –¥–∏–∞–≥–Ω–æ—Å—Ç–∏–∫–∞ –∏—Å—Ç–æ—Ä–∏—á–µ—Å–∫–∏—Ö –¥–∞–Ω–Ω—ã—Ö
# -------------------------------------------------

print("\n1Ô∏è‚É£  –ó–ê–ì–†–£–ó–ö–ê –ò –î–ò–ê–ì–ù–û–°–¢–ò–ö–ê –ò–°–¢–û–†–ò–ß–ï–°–ö–ò–• –î–ê–ù–ù–´–•")
print("-" * 60)

# –ü—É—Ç–∏ –∫ –¥–∞–Ω–Ω—ã–º
# –û–ø—Ä–µ–¥–µ–ª—è–µ–º –±–∞–∑–æ–≤—É—é –¥–∏—Ä–µ–∫—Ç–æ—Ä–∏—é –ø—Ä–æ–µ–∫—Ç–∞ (–∫–æ—Ä–µ–Ω—å –ø—Ä–æ–µ–∫—Ç–∞ finish)
# –ü–æ–¥–Ω–∏–º–∞–µ–º—Å—è –≤–≤–µ—Ä—Ö –ø–æ –¥–∏—Ä–µ–∫—Ç–æ—Ä–∏—è–º, –ø–æ–∫–∞ –Ω–µ –Ω–∞–π–¥–µ–º –ø–∞–ø–∫—É Data
current_dir = os.path.abspath(os.getcwd())
BASE_DIR = current_dir

# –ü–æ–¥–Ω–∏–º–∞–µ–º—Å—è –≤–≤–µ—Ä—Ö, –ø–æ–∫–∞ –Ω–µ –Ω–∞–π–¥–µ–º –ø–∞–ø–∫—É Data –∏–ª–∏ –Ω–µ –¥–æ—Å—Ç–∏–≥–Ω–µ–º –∫–æ—Ä–Ω—è
max_levels = 5  # –ú–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —É—Ä–æ–≤–Ω–µ–π –¥–ª—è –ø–æ–¥—ä–µ–º–∞
for _ in range(max_levels):
    if os.path.exists(os.path.join(BASE_DIR, "Data")):
        break
    parent_dir = os.path.dirname(BASE_DIR)
    if parent_dir == BASE_DIR:  # –î–æ—Å—Ç–∏–≥–ª–∏ –∫–æ—Ä–Ω—è —Ñ–∞–π–ª–æ–≤–æ–π —Å–∏—Å—Ç–µ–º—ã
        break
    BASE_DIR = parent_dir

# –ï—Å–ª–∏ –Ω–µ –Ω–∞—à–ª–∏, –∏—Å–ø–æ–ª—å–∑—É–µ–º —Ñ–∏–∫—Å–∏—Ä–æ–≤–∞–Ω–Ω—ã–π –ø—É—Ç—å: –ø–æ–¥–Ω–∏–º–∞–µ–º—Å—è –Ω–∞ 2 —É—Ä–æ–≤–Ω—è –≤–≤–µ—Ä—Ö
# –æ—Ç —Ç–µ–∫—É—â–µ–π –¥–∏—Ä–µ–∫—Ç–æ—Ä–∏–∏ (notebooks/MiniMax2 -> notebooks -> finish)
if not os.path.exists(os.path.join(BASE_DIR, "Data")):
    BASE_DIR = os.path.abspath(os.path.join(current_dir, os.pardir, os.pardir))

DATA_RAW = os.path.join(BASE_DIR, "Data", "data_raw.csv")
FINAL_XLSX = os.path.join(BASE_DIR, "Data", "final_results_to_analyze.xlsx")

# –í—ã–≤–æ–¥–∏–º –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é –¥–ª—è –æ—Ç–ª–∞–¥–∫–∏
print(f"üìÅ –¢–µ–∫—É—â–∞—è —Ä–∞–±–æ—á–∞—è –¥–∏—Ä–µ–∫—Ç–æ—Ä–∏—è: {os.getcwd()}")
print(f"üìÅ –ë–∞–∑–æ–≤–∞—è –¥–∏—Ä–µ–∫—Ç–æ—Ä–∏—è –ø—Ä–æ–µ–∫—Ç–∞: {BASE_DIR}")
print(f"üìÅ –ü—É—Ç—å –∫ –¥–∞–Ω–Ω—ã–º: {DATA_RAW}")

# –ü—Ä–æ–≤–µ—Ä—è–µ–º –Ω–∞–ª–∏—á–∏–µ —Ñ–∞–π–ª–æ–≤
if not os.path.exists(DATA_RAW):
    raise FileNotFoundError(f"–§–∞–π–ª {DATA_RAW} –Ω–µ –Ω–∞–π–¥–µ–Ω!")
if not os.path.exists(FINAL_XLSX):
    raise FileNotFoundError(f"–§–∞–π–ª {FINAL_XLSX} –Ω–µ –Ω–∞–π–¥–µ–Ω!")

# –ó–∞–≥—Ä—É–∂–∞–µ–º –∏—Å—Ç–æ—Ä–∏—á–µ—Å–∫–∏–µ –¥–∞–Ω–Ω—ã–µ
try:
    raw_data = pd.read_csv(DATA_RAW)
    print(f"‚úÖ –ó–∞–≥—Ä—É–∂–µ–Ω data_raw.csv: {raw_data.shape[0]:,} —Å—Ç—Ä–æ–∫, {raw_data.shape[1]} —Å—Ç–æ–ª–±—Ü–æ–≤")
except Exception as e:
    raise Exception(f"–û—à–∏–±–∫–∞ –∑–∞–≥—Ä—É–∑–∫–∏ data_raw.csv: {e}")

# –î–∏–∞–≥–Ω–æ—Å—Ç–∏–∫–∞ —Å—Ç—Ä—É–∫—Ç—É—Ä—ã –¥–∞–Ω–Ω—ã—Ö
print(f"\nüìã –°—Ç—Ä—É–∫—Ç—É—Ä–∞ –¥–∞–Ω–Ω—ã—Ö:")
print(f"   ‚Ä¢ –°—Ç–æ–ª–±—Ü—ã: {list(raw_data.columns)}")

# –ü—Ä–æ–≤–µ—Ä—è–µ–º –Ω–∞–ª–∏—á–∏–µ –Ω–µ–æ–±—Ö–æ–¥–∏–º—ã—Ö —Å—Ç–æ–ª–±—Ü–æ–≤
required_columns = ['client_id', 'dt', 'event_type']
missing_columns = [col for col in required_columns if col not in raw_data.columns]

if missing_columns:
    print(f"\n‚ö†Ô∏è  –û—Ç—Å—É—Ç—Å—Ç–≤—É—é—Ç —Å—Ç–æ–ª–±—Ü—ã: {missing_columns}")
    print("üîç –ü–æ–∏—Å–∫ –ø–æ—Ö–æ–∂–∏—Ö —Å—Ç–æ–ª–±—Ü–æ–≤...")
    for missing in missing_columns:
        similar = [col for col in raw_data.columns if missing.lower() in col.lower()]
        print(f"   ‚Ä¢ –ü–æ—Ö–æ–∂–∏–µ –Ω–∞ '{missing}': {similar}")
    raise ValueError("–ù–µ–æ–±—Ö–æ–¥–∏–º—ã–µ —Å—Ç–æ–ª–±—Ü—ã –Ω–µ –Ω–∞–π–¥–µ–Ω—ã")

# –ü—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ —Ç–∏–ø–æ–≤ –¥–∞–Ω–Ω—ã—Ö
print(f"\nüîÑ –ü—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ —Ç–∏–ø–æ–≤ –¥–∞–Ω–Ω—ã—Ö...")
raw_data["dt"] = pd.to_datetime(raw_data["dt"])
raw_data["client_id"] = raw_data["client_id"].astype(str)

print(f"   ‚Ä¢ –î–∏–∞–ø–∞–∑–æ–Ω –¥–∞—Ç: {raw_data['dt'].min()} - {raw_data['dt'].max()}")
print(f"   ‚Ä¢ –£–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –∫–ª–∏–µ–Ω—Ç–æ–≤: {raw_data['client_id'].nunique():,}")

# –ê–Ω–∞–ª–∏–∑ —Ç–∏–ø–æ–≤ —Å–æ–±—ã—Ç–∏–π
print(f"\nüìä –ê–Ω–∞–ª–∏–∑ —Å–æ–±—ã—Ç–∏–π:")
event_counts = raw_data['event_type'].value_counts()
print(event_counts)

# –ù–æ—Ä–º–∞–ª–∏–∑–∞—Ü–∏—è –Ω–∞–∑–≤–∞–Ω–∏–π —Å–æ–±—ã—Ç–∏–π
raw_data['event_type'] = raw_data['event_type'].str.strip().str.lower()
event_counts_normalized = raw_data['event_type'].value_counts()
print(f"\nüìä –°–æ–±—ã—Ç–∏—è –ø–æ—Å–ª–µ –Ω–æ—Ä–º–∞–ª–∏–∑–∞—Ü–∏–∏:")
print(event_counts_normalized)

# -------------------------------------------------
# 1.2 –°–æ–∑–¥–∞–Ω–∏–µ –≤–æ—Ä–æ–Ω–∫–∏ —Å–æ–±—ã—Ç–∏–π
# -------------------------------------------------

print(f"\n2Ô∏è‚É£  –ü–û–°–¢–†–û–ï–ù–ò–ï –í–û–†–û–ù–ö–ò –°–û–ë–´–¢–ò–ô")
print("-" * 60)

# –°–æ–∑–¥–∞–Ω–∏–µ –±–∏–Ω–∞—Ä–Ω—ã—Ö –∫–æ–ª–æ–Ω–æ–∫ –Ω–∞ –æ—Å–Ω–æ–≤–µ event_type
# –°–æ–±—ã—Ç–∏—è —É–∂–µ –≤ –ø—Ä–∞–≤–∏–ª—å–Ω–æ–º —Ñ–æ—Ä–º–∞—Ç–µ: 'views', 'clicks', 'adds'
raw_data["views"] = (raw_data['event_type'] == 'views').astype(int)
raw_data["clicks"] = (raw_data['event_type'] == 'clicks').astype(int)
raw_data["adds"] = (raw_data['event_type'] == 'adds').astype(int)

# –ü—Ä–æ–≤–µ—Ä–∫–∞ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤
total_views = raw_data["views"].sum()
total_clicks = raw_data["clicks"].sum()
total_adds = raw_data["adds"].sum()

print(f"\nüìä –ò—Ç–æ–≥–æ–≤—ã–µ —Å–æ–±—ã—Ç–∏—è:")
print(f"   ‚Ä¢ –ü—Ä–æ—Å–º–æ—Ç—Ä—ã: {total_views:,}")
print(f"   ‚Ä¢ –ö–ª–∏–∫–∏: {total_clicks:,}")
print(f"   ‚Ä¢ –î–æ–±–∞–≤–ª–µ–Ω–∏—è: {total_adds:,}")

if total_views == 0 and total_clicks == 0 and total_adds == 0:
    raise ValueError("–í—Å–µ —Å–æ–±—ã—Ç–∏—è –∏–º–µ—é—Ç –Ω—É–ª–µ–≤—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è! –ü—Ä–æ–≤–µ—Ä—å—Ç–µ —Ñ–æ—Ä–º–∞—Ç —Å–æ–±—ã—Ç–∏–π.")

# –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –≤–æ—Ä–æ–Ω–∫–∏ (–Ω–∞–ª–∏—á–∏–µ —Å–æ–±—ã—Ç–∏—è –∑–∞ –¥–∞—Ç—É)
# –ï—Å–ª–∏ —Ö–æ—Ç—è –±—ã —Ä–∞–∑ –Ω–∞ –∫–æ–Ω–∫—Ä–µ—Ç–Ω—É—é –¥–∞—Ç—É –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å –∑–∞—Ö–æ–¥–∏–ª –Ω–∞ –≥–ª–∞–≤–Ω—ã–π —ç–∫—Ä–∞–Ω, —Ç–æ –ø—Ä–∏—Å–≤–∞–∏–≤–∞–µ–º 1
print(f"\nüèóÔ∏è  –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –≤–æ—Ä–æ–Ω–∫–∏...")
funnel = (raw_data.groupby(["client_id", "dt"], as_index=False)
          .agg(views=("views", "max"),
               clicks=("clicks", "max"),
               adds=("adds", "max")))

print(f"   ‚Ä¢ –°—Ç—Ä–æ–∫ –≤ –≤–æ—Ä–æ–Ω–∫–µ: {funnel.shape[0]:,}")
print(f"   ‚Ä¢ –£–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –∫–ª–∏–µ–Ω—Ç–æ–≤: {funnel['client_id'].nunique():,}")

# –ê–≥—Ä–µ–≥–∞—Ü–∏—è –ø–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è–º
user_metrics = (funnel.groupby("client_id", as_index=False)
                .agg(views=("views", "sum"),
                     clicks=("clicks", "sum"),
                     adds=("adds", "sum")))

print(f"   ‚Ä¢ –ü–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –≤ —Ñ–∏–Ω–∞–ª—å–Ω–æ–º –Ω–∞–±–æ—Ä–µ: {user_metrics.shape[0]:,}")

# -------------------------------------------------
# 1.3 –†–∞—Å—á–µ—Ç –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–∏—Ö –º–µ—Ç—Ä–∏–∫
# -------------------------------------------------

print(f"\n3Ô∏è‚É£  –†–ê–°–ß–ï–¢ –ü–û–õ–¨–ó–û–í–ê–¢–ï–õ–¨–°–ö–ò–• –ú–ï–¢–†–ò–ö")
print("-" * 60)

# –ë–∞–∑–æ–≤—ã–µ –º–µ—Ç—Ä–∏–∫–∏
user_metrics["CTR"] = np.where(user_metrics["views"] > 0,
                               user_metrics["clicks"] / user_metrics["views"],
                               np.nan)

user_metrics["CR"] = np.where(user_metrics["clicks"] > 0,
                              user_metrics["adds"] / user_metrics["clicks"],
                              np.nan)

# –°—Ç–∞—Ç–∏—Å—Ç–∏–∫–∏
print(f"üìä –ë–ê–ó–û–í–´–ï –ü–û–ö–ê–ó–ê–¢–ï–õ–ò:")
print(f"   ‚Ä¢ –£–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π: {user_metrics['client_id'].nunique():,}")
print(f"   ‚Ä¢ –í—Å–µ–≥–æ –ø—Ä–æ—Å–º–æ—Ç—Ä–æ–≤: {user_metrics['views'].sum():,}")
print(f"   ‚Ä¢ –í—Å–µ–≥–æ –∫–ª–∏–∫–æ–≤: {user_metrics['clicks'].sum():,}")
print(f"   ‚Ä¢ –í—Å–µ–≥–æ –¥–æ–±–∞–≤–ª–µ–Ω–∏–π: {user_metrics['adds'].sum():,}")
print(f"   ‚Ä¢ –°—Ä–µ–¥–Ω–∏–π CTR: {user_metrics['CTR'].mean():.4f}")
print(f"   ‚Ä¢ –°—Ä–µ–¥–Ω–∏–π CR: {user_metrics['CR'].mean():.4f}")

# –ö–∞—á–µ—Å—Ç–≤–æ –¥–∞–Ω–Ω—ã—Ö
print(f"\nüîç –ö–ê–ß–ï–°–¢–í–û –î–ê–ù–ù–´–•:")
print(f"   ‚Ä¢ –ü–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–∏ —Å –ø—Ä–æ—Å–º–æ—Ç—Ä–∞–º–∏: {(user_metrics['views'] > 0).sum():,}")
print(f"   ‚Ä¢ –ü–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–∏ —Å –∫–ª–∏–∫–∞–º–∏: {(user_metrics['clicks'] > 0).sum():,}")
print(f"   ‚Ä¢ –ü–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–∏ —Å –¥–æ–±–∞–≤–ª–µ–Ω–∏—è–º–∏: {(user_metrics['adds'] > 0).sum():,}")

# -------------------------------------------------
# 1.4 –§—É–Ω–∫—Ü–∏–∏ –¥–ª—è —Ä–∞—Å—á–µ—Ç–æ–≤ (–° –î–ï–õ–¨–¢–ê-–ú–ï–¢–û–î–û–ú)
# -------------------------------------------------

print(f"\n4Ô∏è‚É£  –§–£–ù–ö–¶–ò–ò –î–õ–Ø –†–ê–°–ß–ï–¢–û–í (–° –î–ï–õ–¨–¢–ê-–ú–ï–¢–û–î–û–ú)")
print("-" * 60)

def var_ratio(x, y):
    """
    –î–µ–ª—å—Ç–∞-–º–µ—Ç–æ–¥ –¥–ª—è —Ä–∞—Å—á–µ—Ç–∞ –¥–∏—Å–ø–µ—Ä—Å–∏–∏ ratio-–º–µ—Ç—Ä–∏–∫–∏ x/y
    Var(x/y) ‚âà Var(x)/E(y)¬≤ + Var(y)*E(x)¬≤/E(y)‚Å¥ - 2*E(x)*Cov(x,y)/E(y)¬≥
    """
    mean_x = np.mean(x)
    mean_y = np.mean(y)
    
    # –ü—Ä–æ–≤–µ—Ä—è–µ–º, —á—Ç–æ mean_y –Ω–µ —Ä–∞–≤–µ–Ω –Ω—É–ª—é
    if mean_y == 0:
        return np.inf
    
    var_x = np.var(x, ddof=1)
    var_y = np.var(y, ddof=1)
    cov_xy = np.cov(x, y, ddof=1)[0][1]
    
    # –î–µ–ª—å—Ç–∞-–º–µ—Ç–æ–¥
    variance = (var_x / mean_y**2 
                + var_y * mean_x**2 / mean_y**4 
                - 2 * mean_x * cov_xy / mean_y**3)
    
    return variance

def var_ratio_users(user_df, numerator_col, denominator_col):
    """
    –†–∞—Å—á–µ—Ç –¥–∏—Å–ø–µ—Ä—Å–∏–∏ ratio-–º–µ—Ç—Ä–∏–∫–∏ –Ω–∞ —É—Ä–æ–≤–Ω–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π
    """
    # –§–∏–ª—å—Ç—Ä—É–µ–º –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π, —É –∫–æ—Ç–æ—Ä—ã—Ö denominator > 0
    valid_users = user_df[user_df[denominator_col] > 0].copy()
    
    if len(valid_users) == 0:
        return np.inf
    
    # –†–∞—Å—Å—á–∏—Ç—ã–≤–∞–µ–º ratio –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è
    valid_users['ratio'] = valid_users[numerator_col] / valid_users[denominator_col]
    
    # –í–æ–∑–≤—Ä–∞—â–∞–µ–º –¥–∏—Å–ø–µ—Ä—Å–∏—é —ç—Ç–∏—Ö ratios
    return np.var(valid_users['ratio'], ddof=1)

def get_MDE(mu, std, sample_size, n_groups=2, target_share=0.5, r=1, alpha=0.05, beta=0.2):
    """–†–∞—Å—á–µ—Ç MDE (–º–∏–Ω–∏–º–∞–ª—å–Ω–æ–≥–æ –Ω–∞–±–ª—é–¥–∞–µ–º–æ–≥–æ —ç—Ñ—Ñ–µ–∫—Ç–∞)"""
    t_alpha = stats.norm.ppf(1 - (alpha / 2))
    t_beta = stats.norm.ppf(1 - beta)
    comparisons = n_groups - 1
    sample_ratio_correction = r + 2 + 1/r
    
    mde_abs = (np.sqrt(sample_ratio_correction)*(t_alpha + t_beta) * std / 
               np.sqrt(sample_size*(1-target_share*(comparisons-1))))
    return mde_abs, mde_abs*100/mu

def get_sample_size(mu, var, mde_rel, r=1, alpha=0.05, beta=0.2):
    """–†–∞—Å—á–µ—Ç —Ç—Ä–µ–±—É–µ–º–æ–≥–æ —Ä–∞–∑–º–µ—Ä–∞ –≤—ã–±–æ—Ä–∫–∏"""
    t_alpha = stats.norm.ppf(1 - (alpha / 2))
    t_beta = stats.norm.ppf(1 - beta)
    mde_abs = mu * mde_rel
    
    return ((r + 1) / r) * var * (t_alpha + t_beta) ** 2 / (mde_abs ** 2)

print(f"‚úÖ –§—É–Ω–∫—Ü–∏–∏ –¥–ª—è –¥–µ–ª—å—Ç–∞-–º–µ—Ç–æ–¥–∞ –≥–æ—Ç–æ–≤—ã")

# -------------------------------------------------
# 1.5 –ö–æ—Ä—Ä–µ–∫—Ç–Ω—ã–π —Ä–∞—Å—á–µ—Ç –¥–∏—Å–ø–µ—Ä—Å–∏–π —Å –¥–µ–ª—å—Ç–∞-–º–µ—Ç–æ–¥–æ–º
# -------------------------------------------------

print(f"\n5Ô∏è‚É£  –ö–û–†–†–ï–ö–¢–ù–´–ô –†–ê–°–ß–ï–¢ –î–ò–°–ü–ï–†–°–ò–ô –° –î–ï–õ–¨–¢–ê-–ú–ï–¢–û–î–û–ú")
print("-" * 60)

# –î–ª—è CTR = clicks / views
print(f"üìä –†–ê–°–ß–ï–¢ –î–ò–°–ü–ï–†–°–ò–ò –î–õ–Ø CTR:")
ctr_mean = user_metrics['CTR'].mean()
ctr_var_method1 = var_ratio(user_metrics['clicks'], user_metrics['views'])
ctr_var_method2 = var_ratio_users(user_metrics, 'clicks', 'views')
ctr_std_method1 = np.sqrt(ctr_var_method1)
ctr_std_method2 = np.sqrt(ctr_var_method2)

print(f"   ‚Ä¢ –ú–µ—Ç–æ–¥ 1 (–¥–µ–ª—å—Ç–∞-–º–µ—Ç–æ–¥): Var = {ctr_var_method1:.6f}, Std = {ctr_std_method1:.6f}")
print(f"   ‚Ä¢ –ú–µ—Ç–æ–¥ 2 (–ø–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è–º): Var = {ctr_var_method2:.6f}, Std = {ctr_std_method2:.6f}")
print(f"   ‚Ä¢ –í—ã–±–∏—Ä–∞–µ–º –º–µ—Ç–æ–¥ 1 (–¥–µ–ª—å—Ç–∞-–º–µ—Ç–æ–¥)")
ctr_std = ctr_std_method1

# –î–ª—è CR = adds / clicks (–í–ê–ñ–ù–û - ratio-–º–µ—Ç—Ä–∏–∫–∞!)
print(f"\nüìä –†–ê–°–ß–ï–¢ –î–ò–°–ü–ï–†–°–ò–ò –î–õ–Ø CR (RATIO-–ú–ï–¢–†–ò–ö–ê):")
cr_mean = user_metrics['CR'].mean()
cr_var_method1 = var_ratio(user_metrics['adds'], user_metrics['clicks'])
cr_var_method2 = var_ratio_users(user_metrics, 'adds', 'clicks')
cr_std_method1 = np.sqrt(cr_var_method1)
cr_std_method2 = np.sqrt(cr_var_method2)

print(f"   ‚Ä¢ –ú–µ—Ç–æ–¥ 1 (–¥–µ–ª—å—Ç–∞-–º–µ—Ç–æ–¥): Var = {cr_var_method1:.6f}, Std = {cr_std_method1:.6f}")
print(f"   ‚Ä¢ –ú–µ—Ç–æ–¥ 2 (–ø–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è–º): Var = {cr_var_method2:.6f}, Std = {cr_std_method2:.6f}")
print(f"   ‚Ä¢ –í—ã–±–∏—Ä–∞–µ–º –º–µ—Ç–æ–¥ 1 (–¥–µ–ª—å—Ç–∞-–º–µ—Ç–æ–¥) - –∫–æ—Ä—Ä–µ–∫—Ç–Ω—ã–π –¥–ª—è ratio-–º–µ—Ç—Ä–∏–∫")
cr_std = cr_std_method1

print(f"\nüìã –ò–¢–û–ì–û–í–´–ï –°–¢–ê–¢–ò–°–¢–ò–ö–ò (–° –î–ï–õ–¨–¢–ê-–ú–ï–¢–û–î–û–ú):")
print(f"   ‚Ä¢ CTR: Œº={ctr_mean:.5f}, œÉ={ctr_std:.5f}")
print(f"   ‚Ä¢ CR:  Œº={cr_mean:.5f}, œÉ={cr_std:.5f}")

# -------------------------------------------------
# 1.6 –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–æ–≥–æ –ø–æ–≤–µ–¥–µ–Ω–∏—è
# -------------------------------------------------

print(f"\n6Ô∏è‚É£  –í–ò–ó–£–ê–õ–ò–ó–ê–¶–ò–Ø –ü–û–í–ï–î–ï–ù–ò–Ø –ü–û–õ–¨–ó–û–í–ê–¢–ï–õ–ï–ô (–î–û –≠–ö–°–ü–ï–†–ò–ú–ï–ù–¢–ê)")
print("-" * 60)

# –°–æ–∑–¥–∞–µ–º –≥—Ä–∞—Ñ–∏–∫–∏
fig, axes = plt.subplots(2, 3, figsize=(18, 12))
fig.suptitle('–ê–Ω–∞–ª–∏–∑ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–æ–≥–æ –ø–æ–≤–µ–¥–µ–Ω–∏—è –¥–æ —ç–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç–∞', fontsize=16, fontweight='bold')

# 1. –†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ CTR
axes[0,0].hist(user_metrics["CTR"].dropna(), bins=50, alpha=0.7, color='skyblue', edgecolor='black')
axes[0,0].set_title('–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ CTR')
axes[0,0].set_xlabel('CTR')
axes[0,0].set_ylabel('–ß–∞—Å—Ç–æ—Ç–∞')
axes[0,0].axvline(user_metrics["CTR"].mean(), color='red', linestyle='--', 
                 label=f'–°—Ä–µ–¥–Ω–µ–µ: {user_metrics["CTR"].mean():.4f}')
axes[0,0].legend()

# 2. –†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ CR (–≤–∞–∂–Ω–æ!)
valid_cr = user_metrics["CR"].dropna()
axes[0,1].hist(valid_cr, bins=50, alpha=0.7, color='lightgreen', edgecolor='black')
axes[0,1].set_title('–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ CR (ratio-–º–µ—Ç—Ä–∏–∫–∞)')
axes[0,1].set_xlabel('CR = adds/clicks')
axes[0,1].set_ylabel('–ß–∞—Å—Ç–æ—Ç–∞')
axes[0,1].axvline(valid_cr.mean(), color='red', linestyle='--',
                 label=f'–°—Ä–µ–¥–Ω–µ–µ: {valid_cr.mean():.4f}')
axes[0,1].legend()

# 3. Scatter plot: —Å–æ–æ—Ç–Ω–æ—à–µ–Ω–∏–µ –∫–ª–∏–∫–æ–≤ –∏ –¥–æ–±–∞–≤–ª–µ–Ω–∏–π
valid_users = user_metrics[(user_metrics['clicks'] > 0) & (user_metrics['adds'] >= 0)]
if len(valid_users) > 0:
    sample_size = min(5000, len(valid_users))
    sample_data = valid_users.sample(sample_size)
    axes[0,2].scatter(sample_data['clicks'], sample_data['adds'], alpha=0.5, s=1)
    axes[0,2].set_title('–°–æ–æ—Ç–Ω–æ—à–µ–Ω–∏–µ –∫–ª–∏–∫–æ–≤ –∏ –¥–æ–±–∞–≤–ª–µ–Ω–∏–π')
    axes[0,2].set_xlabel('–ö–ª–∏–∫–∏')
    axes[0,2].set_ylabel('–î–æ–±–∞–≤–ª–µ–Ω–∏—è')

# 4. –ö–æ—Ä—Ä–µ–ª—è—Ü–∏—è –º–µ–∂–¥—É –º–µ—Ç—Ä–∏–∫–∞–º–∏
corr_data = user_metrics[['views', 'clicks', 'adds']].corr()
sns.heatmap(corr_data, annot=True, cmap='coolwarm', center=0, ax=axes[1,0])
axes[1,0].set_title('–ö–æ—Ä—Ä–µ–ª—è—Ü–∏—è –º–µ–∂–¥—É –º–µ—Ç—Ä–∏–∫–∞–º–∏')

# 5. –í—Ä–µ–º–µ–Ω–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Å–æ–±—ã—Ç–∏–π
daily_events = raw_data.groupby(raw_data['dt'].dt.date).agg({
    'views': 'sum',
    'clicks': 'sum', 
    'adds': 'sum'
}).reset_index()

axes[1,1].plot(daily_events['dt'], daily_events['views'], label='–ü—Ä–æ—Å–º–æ—Ç—Ä—ã', marker='o')
axes[1,1].plot(daily_events['dt'], daily_events['clicks'], label='–ö–ª–∏–∫–∏', marker='s')
axes[1,1].plot(daily_events['dt'], daily_events['adds'], label='–î–æ–±–∞–≤–ª–µ–Ω–∏—è', marker='^')
axes[1,1].set_title('–í—Ä–µ–º–µ–Ω–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Å–æ–±—ã—Ç–∏–π')
axes[1,1].set_xlabel('–î–∞—Ç–∞')
axes[1,1].set_ylabel('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å–æ–±—ã—Ç–∏–π')
axes[1,1].legend()
axes[1,1].tick_params(axis='x', rotation=45)

# 6. –†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ CR –ø–æ –∫–≤–∞–Ω—Ç–∏–ª—è–º –∫–ª–∏–∫–æ–≤
if len(valid_users) > 0:
    try:
        # –ü—ã—Ç–∞–µ–º—Å—è –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å qcut —Å –æ–±—Ä–∞–±–æ—Ç–∫–æ–π –¥—É–±–ª–∏–∫–∞—Ç–æ–≤
        valid_users['clicks_quantile'] = pd.qcut(valid_users['clicks'], q=5, labels=['Q1', 'Q2', 'Q3', 'Q4', 'Q5'], duplicates='drop')
        cr_by_quantile = valid_users.groupby('clicks_quantile')['CR'].mean()
        axes[1,2].bar(cr_by_quantile.index, cr_by_quantile.values, alpha=0.7, color='orange')
        axes[1,2].set_title('CR –ø–æ –∫–≤–∞–Ω—Ç–∏–ª—è–º –∫–ª–∏–∫–æ–≤')
        axes[1,2].set_xlabel('–ö–≤–∞–Ω—Ç–∏–ª—å –∫–ª–∏–∫–æ–≤')
        axes[1,2].set_ylabel('–°—Ä–µ–¥–Ω–∏–π CR')
    except ValueError:
        # –ï—Å–ª–∏ qcut –Ω–µ —Ä–∞–±–æ—Ç–∞–µ—Ç –∏–∑-–∑–∞ –¥—É–±–ª–∏–∫–∞—Ç–æ–≤, –∏—Å–ø–æ–ª—å–∑—É–µ–º cut —Å —Ä–∞–≤–Ω—ã–º–∏ –∏–Ω—Ç–µ—Ä–≤–∞–ª–∞–º–∏
        try:
            valid_users['clicks_quantile'] = pd.cut(valid_users['clicks'], bins=5, labels=['Q1', 'Q2', 'Q3', 'Q4', 'Q5'], duplicates='drop')
            cr_by_quantile = valid_users.groupby('clicks_quantile')['CR'].mean()
            axes[1,2].bar(cr_by_quantile.index, cr_by_quantile.values, alpha=0.7, color='orange')
            axes[1,2].set_title('CR –ø–æ –∏–Ω—Ç–µ—Ä–≤–∞–ª–∞–º –∫–ª–∏–∫–æ–≤')
            axes[1,2].set_xlabel('–ò–Ω—Ç–µ—Ä–≤–∞–ª –∫–ª–∏–∫–æ–≤')
            axes[1,2].set_ylabel('–°—Ä–µ–¥–Ω–∏–π CR')
        except:
            # –ï—Å–ª–∏ –∏ —ç—Ç–æ –Ω–µ —Ä–∞–±–æ—Ç–∞–µ—Ç, –∏—Å–ø–æ–ª—å–∑—É–µ–º –ø—Ä–æ—Å—Ç—É—é –≥—Ä—É–ø–ø–∏—Ä–æ–≤–∫—É –ø–æ –∑–Ω–∞—á–µ–Ω–∏—è–º
            unique_clicks = sorted(valid_users['clicks'].unique())
            if len(unique_clicks) >= 5:
                # –†–∞–∑–±–∏–≤–∞–µ–º –Ω–∞ 5 –≥—Ä—É–ø–ø –ø–æ –∑–Ω–∞—á–µ–Ω–∏—è–º
                click_bins = np.linspace(valid_users['clicks'].min(), valid_users['clicks'].max(), 6)
                valid_users['clicks_quantile'] = pd.cut(valid_users['clicks'], bins=click_bins, labels=['Q1', 'Q2', 'Q3', 'Q4', 'Q5'], include_lowest=True, duplicates='drop')
                cr_by_quantile = valid_users.groupby('clicks_quantile')['CR'].mean()
                axes[1,2].bar(range(len(cr_by_quantile)), cr_by_quantile.values, alpha=0.7, color='orange')
                axes[1,2].set_xticks(range(len(cr_by_quantile)))
                axes[1,2].set_xticklabels(cr_by_quantile.index)
                axes[1,2].set_title('CR –ø–æ –∏–Ω—Ç–µ—Ä–≤–∞–ª–∞–º –∫–ª–∏–∫–æ–≤')
                axes[1,2].set_xlabel('–ò–Ω—Ç–µ—Ä–≤–∞–ª –∫–ª–∏–∫–æ–≤')
                axes[1,2].set_ylabel('–°—Ä–µ–¥–Ω–∏–π CR')
            else:
                # –ï—Å–ª–∏ —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π –º–µ–Ω—å—à–µ 5, –ø—Ä–æ—Å—Ç–æ –ø–æ–∫–∞–∑—ã–≤–∞–µ–º –ø–æ –∫–∞–∂–¥–æ–º—É –∑–Ω–∞—á–µ–Ω–∏—é
                cr_by_clicks = valid_users.groupby('clicks')['CR'].mean()
                axes[1,2].bar(range(len(cr_by_clicks)), cr_by_clicks.values, alpha=0.7, color='orange')
                axes[1,2].set_xticks(range(len(cr_by_clicks)))
                axes[1,2].set_xticklabels(cr_by_clicks.index, rotation=45)
                axes[1,2].set_title('CR –ø–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤—É –∫–ª–∏–∫–æ–≤')
                axes[1,2].set_xlabel('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –∫–ª–∏–∫–æ–≤')
                axes[1,2].set_ylabel('–°—Ä–µ–¥–Ω–∏–π CR')

plt.tight_layout()
plt.savefig('user_behavior_before_experiment.png', dpi=300, bbox_inches='tight')
plt.show()

# -------------------------------------------------
# 1.7 –í—ã–≤–æ–¥—ã –æ –ø–æ–≤–µ–¥–µ–Ω–∏–∏ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π
# -------------------------------------------------

print(f"\n7Ô∏è‚É£  –í–´–í–û–î–´ –û –ü–û–í–ï–î–ï–ù–ò–ò –ü–û–õ–¨–ó–û–í–ê–¢–ï–õ–ï–ô")
print("-" * 60)

# –ê–Ω–∞–ª–∏–∑ —Å–µ–≥–º–µ–Ω—Ç–æ–≤
clickers = user_metrics[user_metrics['clicks'] > 0]
non_clickers = user_metrics[user_metrics['clicks'] == 0]

print(f"üìä –ê–ù–ê–õ–ò–ó –°–ï–ì–ú–ï–ù–¢–û–í:")
print(f"   ‚Ä¢ –ü–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–∏ —Å –∫–ª–∏–∫–∞–º–∏: {len(clickers):,} ({len(clickers)/len(user_metrics):.1%})")
print(f"   ‚Ä¢ –ü–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–∏ –±–µ–∑ –∫–ª–∏–∫–æ–≤: {len(non_clickers):,} ({len(non_clickers)/len(user_metrics):.1%})")

if len(clickers) > 0:
    print(f"   ‚Ä¢ –°—Ä–µ–¥–Ω–∏–π CTR —É –∞–∫—Ç–∏–≤–Ω—ã—Ö –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π: {clickers['CTR'].mean():.4f}")
    print(f"   ‚Ä¢ –°—Ä–µ–¥–Ω–∏–π CR —É –∫–ª–∏–∫–∞–≤—à–∏—Ö: {clickers['CR'].mean():.4f}")
    
    # –ö–æ–Ω–≤–µ—Ä—Å–∏—è –≤ –¥–æ–±–∞–≤–ª–µ–Ω–∏–µ —Å—Ä–µ–¥–∏ –∫–ª–∏–∫–∞–≤—à–∏—Ö
    adds_from_clicks = clickers['adds'].sum() / clickers['clicks'].sum()
    print(f"   ‚Ä¢ –û–±—â–∞—è CR (—Å—Ä–µ–¥–∏ –∫–ª–∏–∫–∞–≤—à–∏—Ö): {adds_from_clicks:.4f}")

print(f"\nüìù –í–´–í–û–î–´ –û CR (RATIO-–ú–ï–¢–†–ò–ö–ï):")
print(f"   ‚Ä¢ CR –ø–æ–∫–∞–∑—ã–≤–∞–µ—Ç —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Å –¥–ª–∏–Ω–Ω—ã–º —Ö–≤–æ—Å—Ç–æ–º")
print(f"   ‚Ä¢ –ü—Ä–∏–º–µ–Ω–µ–Ω–∏–µ –¥–µ–ª—å—Ç–∞-–º–µ—Ç–æ–¥–∞ –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ –¥–ª—è –∫–æ—Ä—Ä–µ–∫—Ç–Ω–æ–≥–æ —Ä–∞—Å—á–µ—Ç–∞ –¥–∏—Å–ø–µ—Ä—Å–∏–∏")
print(f"   ‚Ä¢ –°—Ä–µ–¥–Ω–∏–π CR: {cr_mean:.4f}, —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ: {cr_std:.5f}")
print(f"   ‚Ä¢ –ö–æ—Ä—Ä–µ–ª—è—Ü–∏—è –º–µ–∂–¥—É –∫–ª–∏–∫–∞–º–∏ –∏ –¥–æ–±–∞–≤–ª–µ–Ω–∏—è–º–∏: {np.corrcoef(clickers['clicks'], clickers['adds'])[0,1]:.3f}")

# -------------------------------------------------
# 1.8 –†–∞—Å—á–µ—Ç MDE –∏ —Ä–∞–∑–º–µ—Ä–∞ –≤—ã–±–æ—Ä–∫–∏ (–° –î–ï–õ–¨–¢–ê-–ú–ï–¢–û–î–û–ú)
# -------------------------------------------------

print(f"\n8Ô∏è‚É£  –†–ê–°–ß–ï–¢ MDE –ò –†–ê–ó–ú–ï–†–ê –í–´–ë–û–†–ö–ò (–° –î–ï–õ–¨–¢–ê-–ú–ï–¢–û–î–û–ú)")
print("-" * 60)

# –û–≥—Ä–∞–Ω–∏—á–µ–Ω–∏—è —ç–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç–∞
total_users = user_metrics["client_id"].nunique()
max_total_sample = int(0.20 * total_users)  # 20% –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π
target_share = 0.5
sample_per_group = int(max_total_sample * target_share)

print(f"üîí –û–ì–†–ê–ù–ò–ß–ï–ù–ò–Ø –≠–ö–°–ü–ï–†–ò–ú–ï–ù–¢–ê:")
print(f"   ‚Ä¢ –í—Å–µ–≥–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π: {total_users:,}")
print(f"   ‚Ä¢ –ú–∞–∫—Å–∏–º–∞–ª—å–Ω—ã–π —Ä–∞–∑–º–µ—Ä —ç–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç–∞ (20%): {max_total_sample:,}")
print(f"   ‚Ä¢ –†–∞–∑–º–µ—Ä –∫–∞–∂–¥–æ–π –≥—Ä—É–ø–ø—ã (50/50): {sample_per_group:,}")

# –¢–∞–±–ª–∏—Ü–∞ MDE –¥–ª—è —Ä–∞–∑–Ω—ã—Ö –≤–∞—Ä–∏–∞–Ω—Ç–æ–≤
mde_values = [0.02, 0.03, 0.04, 0.05, 0.10]
mde_results = []

print(f"\nüìä –†–ê–°–ß–ï–¢ MDE –î–õ–Ø CTR:")
for mde in mde_values:
    sample_needed = get_sample_size(ctr_mean, ctr_std**2, mde, alpha=0.05, beta=0.20)
    mde_abs, mde_rel_calc = get_MDE(ctr_mean, ctr_std, sample_per_group * 2, alpha=0.05, beta=0.20)
    
    mde_results.append({
        'MDE_—Ü–µ–ª–µ–≤–æ–π': f"{mde:.1%}",
        'MDE_–¥–æ—Å—Ç–∏–∂–∏–º—ã–π': f"{mde_rel_calc:.2%}",
        '–†–∞–∑–º–µ—Ä_–≥—Ä—É–ø–ø—ã_–Ω—É–∂–µ–Ω': f"{sample_needed:,.0f}",
        'CTR_–±–∞–∑–æ–≤—ã–π': f"{ctr_mean:.4f}",
        'CTR_std': f"{ctr_std:.5f}",
        '–î–æ—Å—Ç—É–ø–Ω–æ': '‚úÖ' if sample_needed <= sample_per_group else '‚ùå'
    })

print(f"\nüìä –†–ê–°–ß–ï–¢ MDE –î–õ–Ø CR (RATIO-–ú–ï–¢–†–ò–ö–ê):")
cr_mde_results = []
for mde in mde_values:
    sample_needed = get_sample_size(cr_mean, cr_std**2, mde, alpha=0.05, beta=0.20)
    mde_abs, mde_rel_calc = get_MDE(cr_mean, cr_std, sample_per_group * 2, alpha=0.05, beta=0.20)
    
    cr_mde_results.append({
        'MDE_—Ü–µ–ª–µ–≤–æ–π': f"{mde:.1%}",
        'MDE_–¥–æ—Å—Ç–∏–∂–∏–º—ã–π': f"{mde_rel_calc:.2%}",
        '–†–∞–∑–º–µ—Ä_–≥—Ä—É–ø–ø—ã_–Ω—É–∂–µ–Ω': f"{sample_needed:,.0f}",
        'CR_–±–∞–∑–æ–≤—ã–π': f"{cr_mean:.4f}",
        'CR_std': f"{cr_std:.5f}",
        '–î–æ—Å—Ç—É–ø–Ω–æ': '‚úÖ' if sample_needed <= sample_per_group else '‚ùå'
    })

mde_df = pd.DataFrame(mde_results)
cr_mde_df = pd.DataFrame(cr_mde_results)

print(f"\nüìã –¢–ê–ë–õ–ò–¶–ê MDE –î–õ–Ø CTR:")
print(mde_df.to_string(index=False))

print(f"\nüìã –¢–ê–ë–õ–ò–¶–ê MDE –î–õ–Ø CR (RATIO-–ú–ï–¢–†–ò–ö–ê):")
print(cr_mde_df.to_string(index=False))

# –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è MDE
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(16, 12))
fig.suptitle('–ê–Ω–∞–ª–∏–∑ MDE —Å –¥–µ–ª—å—Ç–∞-–º–µ—Ç–æ–¥–æ–º', fontsize=16, fontweight='bold')

# CTR MDE
achievable_mdes_ctr = [float(row['MDE_–¥–æ—Å—Ç–∏–∂–∏–º—ã–π'].rstrip('%')) for row in mde_results]
ax1.bar(range(len(achievable_mdes_ctr)), achievable_mdes_ctr, color='skyblue', alpha=0.7)
ax1.set_title('–î–æ—Å—Ç–∏–∂–∏–º—ã–π MDE –¥–ª—è CTR')
ax1.set_xlabel('–í–∞—Ä–∏–∞–Ω—Ç—ã MDE')
ax1.set_ylabel('MDE (%)')
ax1.set_xticks(range(len(mde_values)))
ax1.set_xticklabels([f"{mde:.1%}" for mde in mde_values])

# CTR —Ä–∞–∑–º–µ—Ä –≤—ã–±–æ—Ä–∫–∏
required_sizes_ctr = [float(row['–†–∞–∑–º–µ—Ä_–≥—Ä—É–ø–ø—ã_–Ω—É–∂–µ–Ω'].replace(',', '')) for row in mde_results]
ax2.bar(range(len(required_sizes_ctr)), required_sizes_ctr, color='lightcoral', alpha=0.7)
ax2.axhline(y=sample_per_group, color='red', linestyle='--', label=f'–î–æ—Å—Ç—É–ø–Ω–æ: {sample_per_group:,}')
ax2.set_title('–¢—Ä–µ–±—É–µ–º—ã–π —Ä–∞–∑–º–µ—Ä –≥—Ä—É–ø–ø—ã –¥–ª—è CTR')
ax2.set_xlabel('–í–∞—Ä–∏–∞–Ω—Ç—ã MDE')
ax2.set_ylabel('–†–∞–∑–º–µ—Ä –≥—Ä—É–ø–ø—ã')
ax2.set_xticks(range(len(mde_values)))
ax2.set_xticklabels([f"{mde:.1%}" for mde in mde_values])
ax2.legend()

# CR MDE
achievable_mdes_cr = [float(row['MDE_–¥–æ—Å—Ç–∏–∂–∏–º—ã–π'].rstrip('%')) for row in cr_mde_results]
ax3.bar(range(len(achievable_mdes_cr)), achievable_mdes_cr, color='lightgreen', alpha=0.7)
ax3.set_title('–î–æ—Å—Ç–∏–∂–∏–º—ã–π MDE –¥–ª—è CR (ratio-–º–µ—Ç—Ä–∏–∫–∞)')
ax3.set_xlabel('–í–∞—Ä–∏–∞–Ω—Ç—ã MDE')
ax3.set_ylabel('MDE (%)')
ax3.set_xticks(range(len(mde_values)))
ax3.set_xticklabels([f"{mde:.1%}" for mde in mde_values])

# CR —Ä–∞–∑–º–µ—Ä –≤—ã–±–æ—Ä–∫–∏
required_sizes_cr = [float(row['–†–∞–∑–º–µ—Ä_–≥—Ä—É–ø–ø—ã_–Ω—É–∂–µ–Ω'].replace(',', '')) for row in cr_mde_results]
ax4.bar(range(len(required_sizes_cr)), required_sizes_cr, color='orange', alpha=0.7)
ax4.axhline(y=sample_per_group, color='red', linestyle='--', label=f'–î–æ—Å—Ç—É–ø–Ω–æ: {sample_per_group:,}')
ax4.set_title('–¢—Ä–µ–±—É–µ–º—ã–π —Ä–∞–∑–º–µ—Ä –≥—Ä—É–ø–ø—ã –¥–ª—è CR')
ax4.set_xlabel('–í–∞—Ä–∏–∞–Ω—Ç—ã MDE')
ax4.set_ylabel('–†–∞–∑–º–µ—Ä –≥—Ä—É–ø–ø—ã')
ax4.set_xticks(range(len(mde_values)))
ax4.set_xticklabels([f"{mde:.1%}" for mde in mde_values])
ax4.legend()

plt.tight_layout()
plt.savefig('mde_analysis_with_delta_method.png', dpi=300, bbox_inches='tight')
plt.show()

# -------------------------------------------------
# 1.9 –í—ã–±–æ—Ä –æ–ø—Ç–∏–º–∞–ª—å–Ω–æ–≥–æ MDE
# -------------------------------------------------

print(f"\n9Ô∏è‚É£  –í–´–ë–û–† –û–ü–¢–ò–ú–ê–õ–¨–ù–û–ì–û MDE")
print("-" * 60)

# –ê–Ω–∞–ª–∏–∑ –¥–æ—Å—Ç–∏–∂–∏–º—ã—Ö MDE –¥–ª—è CTR
achievable_mdes_ctr = []
for mde in mde_values:
    sample_needed = get_sample_size(ctr_mean, ctr_std**2, mde, alpha=0.05, beta=0.20)
    if sample_needed <= sample_per_group:
        achievable_mdes_ctr.append(mde)

# –ê–Ω–∞–ª–∏–∑ –¥–æ—Å—Ç–∏–∂–∏–º—ã—Ö MDE –¥–ª—è CR
achievable_mdes_cr = []
for mde in mde_values:
    sample_needed = get_sample_size(cr_mean, cr_std**2, mde, alpha=0.05, beta=0.20)
    if sample_needed <= sample_per_group:
        achievable_mdes_cr.append(mde)

print(f"üìä –ê–ù–ê–õ–ò–ó –î–û–°–¢–ò–ñ–ò–ú–´–• MDE:")
print(f"   ‚Ä¢ CTR - –¥–æ—Å—Ç–∏–∂–∏–º—ã–µ MDE: {[f'{mde:.1%}' for mde in achievable_mdes_ctr]}")
print(f"   ‚Ä¢ CR - –¥–æ—Å—Ç–∏–∂–∏–º—ã–µ MDE: {[f'{mde:.1%}' for mde in achievable_mdes_cr]}")

if achievable_mdes_ctr and achievable_mdes_cr:
    recommended_mde_ctr = max(achievable_mdes_ctr)
    recommended_mde_cr = max(achievable_mdes_cr)
    
    print(f"\n‚úÖ –†–ï–ö–û–ú–ï–ù–î–£–ï–ú–´–ï MDE:")
    print(f"   ‚Ä¢ CTR: {recommended_mde_ctr:.1%}")
    print(f"   ‚Ä¢ CR: {recommended_mde_cr:.1%}")
    
    # –ò—Å–ø–æ–ª—å–∑—É–µ–º –º–∏–Ω–∏–º–∞–ª—å–Ω—ã–π –∏–∑ –Ω–∏—Ö –¥–ª—è –∫–æ–Ω—Å–µ—Ä–≤–∞—Ç–∏–≤–Ω–æ—Å—Ç–∏
    recommended_mde = min(recommended_mde_ctr, recommended_mde_cr)
    print(f"   ‚Ä¢ –§–∏–Ω–∞–ª—å–Ω—ã–π MDE (–∫–æ–Ω—Å–µ—Ä–≤–∞—Ç–∏–≤–Ω—ã–π): {recommended_mde:.1%}")
    
    sample_needed = get_sample_size(ctr_mean, ctr_std**2, recommended_mde, alpha=0.05, beta=0.20)
    
    print(f"   ‚Ä¢ –¢—Ä–µ–±—É–µ–º—ã–π —Ä–∞–∑–º–µ—Ä –≥—Ä—É–ø–ø—ã: {sample_needed:,.0f}")
    print(f"   ‚Ä¢ –î–æ—Å—Ç—É–ø–Ω—ã–π —Ä–∞–∑–º–µ—Ä –≥—Ä—É–ø–ø—ã: {sample_per_group:,}")
    print(f"   ‚Ä¢ –ó–∞–ø–∞—Å –º–æ—â–Ω–æ—Å—Ç–∏: {(sample_per_group - sample_needed)/sample_needed:.1%}")
    
    # –°–æ—Ö—Ä–∞–Ω–µ–Ω–∏–µ –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ —ç–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç–∞
    experiment_params = {
        'mde': recommended_mde,
        'sample_per_group': sample_per_group,
        'total_users': total_users,
        'ctr_baseline': ctr_mean,
        'cr_baseline': cr_mean,
        'ctr_std': ctr_std,
        'cr_std': cr_std,
        'alpha': 0.05,
        'beta': 0.20,
        'method': 'delta_method'
    }
    
    print(f"\nüìã –ü–ê–†–ê–ú–ï–¢–†–´ –≠–ö–°–ü–ï–†–ò–ú–ï–ù–¢–ê:")
    for key, value in experiment_params.items():
        if isinstance(value, float):
            print(f"   ‚Ä¢ {key}: {value:.5f}")
        elif isinstance(value, (int, np.integer)):
            print(f"   ‚Ä¢ {key}: {value:,}")
        else:
            print(f"   ‚Ä¢ {key}: {value}")
            
else:
    print("‚ùå –í–ù–ò–ú–ê–ù–ò–ï: –ù–∏ –æ–¥–∏–Ω –∏–∑ –ø–ª–∞–Ω–∏—Ä—É–µ–º—ã—Ö MDE –Ω–µ –¥–æ—Å—Ç–∏–∂–∏–º –ø—Ä–∏ –æ–≥—Ä–∞–Ω–∏—á–µ–Ω–∏–∏ 20% –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π")
    experiment_params = None

# ===================================================================
# –ß–ê–°–¢–¨ 2: –ê–ù–ê–õ–ò–ó –†–ï–ó–£–õ–¨–¢–ê–¢–û–í –≠–ö–°–ü–ï–†–ò–ú–ï–ù–¢–ê
# ===================================================================

print("\n" + "="*80)
print("üî¨ –ß–ê–°–¢–¨ 2: –ê–ù–ê–õ–ò–ó –†–ï–ó–£–õ–¨–¢–ê–¢–û–í –≠–ö–°–ü–ï–†–ò–ú–ï–ù–¢–ê")
print("="*80)

# -------------------------------------------------
# 2.1 –ó–∞–≥—Ä—É–∑–∫–∞ –∏ –ø–æ–¥–≥–æ—Ç–æ–≤–∫–∞ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤ —ç–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç–∞
# -------------------------------------------------

print(f"\n1Ô∏è‚É£  –ó–ê–ì–†–£–ó–ö–ê –ò –ü–û–î–ì–û–¢–û–í–ö–ê –†–ï–ó–£–õ–¨–¢–ê–¢–û–í –≠–ö–°–ü–ï–†–ò–ú–ï–ù–¢–ê")
print("-" * 60)

# –ó–∞–≥—Ä—É–∑–∫–∞ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤
try:
    results_data = pd.read_excel(FINAL_XLSX)
    print(f"‚úÖ –ó–∞–≥—Ä—É–∂–µ–Ω final_results_to_analyze.xlsx: {results_data.shape[0]:,} —Å—Ç—Ä–æ–∫")
except Exception as e:
    raise Exception(f"–û—à–∏–±–∫–∞ –∑–∞–≥—Ä—É–∑–∫–∏ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤: {e}")

# –î–∏–∞–≥–Ω–æ—Å—Ç–∏–∫–∞ —Å—Ç—Ä—É–∫—Ç—É—Ä—ã
print(f"\nüìã –°—Ç—Ä—É–∫—Ç—É—Ä–∞ —Ñ–∞–π–ª–∞ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤:")
print(f"   ‚Ä¢ –°—Ç–æ–ª–±—Ü—ã: {list(results_data.columns)}")
print(f"   ‚Ä¢ –ü–µ—Ä–∏–æ–¥: {results_data['date'].min()} - {results_data['date'].max()}")

# –ü–æ–∏—Å–∫ —Å—Ç–æ–ª–±—Ü–∞ —Å –≥—Ä—É–ø–ø–∞–º–∏
possible_group_columns = ['ab_group', 'group', 'variant', 'test_group', '–ì—Ä—É–ø–ø–∞', '–í–∞—Ä–∏–∞–Ω—Ç']
group_column = None

for col in possible_group_columns:
    if col in results_data.columns:
        group_column = col
        break

if group_column is None:
    # –ò—â–µ–º —Å—Ç–æ–ª–±–µ—Ü —Å 2-3 —É–Ω–∏–∫–∞–ª—å–Ω—ã–º–∏ –∑–Ω–∞—á–µ–Ω–∏—è–º–∏
    for col in results_data.columns:
        unique_count = results_data[col].nunique()
        if 2 <= unique_count <= 3:
            group_column = col
            break

if group_column is None:
    raise ValueError("–ù–µ —É–¥–∞–ª–æ—Å—å –Ω–∞–π—Ç–∏ —Å—Ç–æ–ª–±–µ—Ü —Å –≥—Ä—É–ø–ø–∞–º–∏ —ç–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç–∞")

print(f"‚úÖ –ù–∞–π–¥–µ–Ω —Å—Ç–æ–ª–±–µ—Ü –≥—Ä—É–ø–ø: {group_column}")

# –ü–æ–ª—É—á–∞–µ–º –æ—Å–Ω–æ–≤–Ω—ã–µ –≥—Ä—É–ø–ø—ã
group_counts = results_data[group_column].value_counts()
main_groups = group_counts.head(2).index.tolist()

print(f"   ‚Ä¢ –û—Å–Ω–æ–≤–Ω—ã–µ –≥—Ä—É–ø–ø—ã: {main_groups}")
print(f"   ‚Ä¢ –†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ:")
for group, count in group_counts.items():
    print(f"     - {group}: {count:,} –∑–∞–ø–∏—Å–µ–π ({count/len(results_data):.1%})")

# –§–∏–ª—å—Ç—Ä–∞—Ü–∏—è –∏ –ø–æ–¥–≥–æ—Ç–æ–≤–∫–∞ –¥–∞–Ω–Ω—ã—Ö
clean_results = results_data[results_data[group_column].isin(main_groups)].copy()
clean_results['ab_group'] = clean_results[group_column]

# –£–¥–∞–ª–µ–Ω–∏–µ –¥—É–±–ª–∏–∫–∞—Ç–æ–≤
initial_rows = len(clean_results)
clean_results = clean_results.drop_duplicates(subset=['client_id', 'date', 'ab_group'])
final_rows = len(clean_results)

print(f"   ‚Ä¢ –ü–æ—Å–ª–µ —É–¥–∞–ª–µ–Ω–∏—è –¥—É–±–ª–∏–∫–∞—Ç–æ–≤: {final_rows:,} –∑–∞–ø–∏—Å–µ–π (—É–¥–∞–ª–µ–Ω–æ {initial_rows - final_rows:,})")

# -------------------------------------------------
# 2.2 –°—Ä–∞–≤–Ω–µ–Ω–∏–µ –∫–ª—é—á–µ–≤—ã—Ö –º–µ—Ç—Ä–∏–∫ –º–µ–∂–¥—É –≥—Ä—É–ø–ø–∞–º–∏
# -------------------------------------------------

print(f"\n2Ô∏è‚É£  –°–†–ê–í–ù–ï–ù–ò–ï –ö–õ–Æ–ß–ï–í–´–• –ú–ï–¢–†–ò–ö –ú–ï–ñ–î–£ –ì–†–£–ü–ü–ê–ú–ò")
print("-" * 60)

# –†–∞—Å—á–µ—Ç –º–µ—Ç—Ä–∏–∫ –ø–æ –≥—Ä—É–ø–ø–∞–º
def calculate_group_metrics(data, group_name):
    """–†–∞—Å—á–µ—Ç –º–µ—Ç—Ä–∏–∫ –¥–ª—è –≥—Ä—É–ø–ø—ã"""
    group_data = data[data['ab_group'] == group_name]
    
    # cnt_view_ads - —ç—Ç–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø—Ä–æ—Å–º–æ—Ç—Ä–æ–≤ —Ä–µ–∫–ª–∞–º—ã (–∫–ª–∏–∫–∏ –Ω–∞ —Ä–µ–∫–ª–∞–º—É)
    # cnt_adds_ads - –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –¥–æ–±–∞–≤–ª–µ–Ω–∏–π –∏–∑ —Ä–µ–∫–ª–∞–º—ã
    total_clicks = group_data['cnt_view_ads'].sum()
    total_adds = group_data['cnt_adds_ads'].sum()
    total_revenue = group_data['sum_adds_ads'].sum()
    total_orders = group_data['cnt_orders_ads'].sum()
    total_orders_revenue = group_data['sum_orders_ads'].sum()
    
    # –ü–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–∏ —Å –ø—Ä–æ—Å–º–æ—Ç—Ä–∞–º–∏ —Ä–µ–∫–ª–∞–º—ã (–∫–ª–∏–∫–∞–º–∏)
    users_with_clicks = group_data[group_data['cnt_view_ads'] > 0]['client_id'].nunique() if len(group_data) > 0 else 0
    # –ü–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–∏ —Å –¥–æ–±–∞–≤–ª–µ–Ω–∏—è–º–∏
    users_with_adds = group_data[group_data['cnt_adds_ads'] > 0]['client_id'].nunique() if len(group_data) > 0 else 0
    # –í—Å–µ–≥–æ —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π
    total_users = group_data['client_id'].nunique()
    
    metrics = {
        'users': total_users,
        'users_with_clicks': users_with_clicks,
        'users_with_adds': users_with_adds,
        'records': len(group_data),
        'total_clicks': total_clicks,
        'total_adds': total_adds,
        'total_revenue': total_revenue,
        'total_orders': total_orders,
        'total_orders_revenue': total_orders_revenue
    }
    
    # –ü—Ä–æ–∏–∑–≤–æ–¥–Ω—ã–µ –º–µ—Ç—Ä–∏–∫–∏
    # CTR = –∫–ª–∏–∫–∏ / –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–∏ (–∫–ª–∏–∫–∞–±–µ–ª—å–Ω–æ—Å—Ç—å —Å–µ–∫—Ü–∏–∏ —Ä–µ–∫–ª–∞–º—ã)
    if metrics['users'] > 0:
        metrics['ctr'] = metrics['total_clicks'] / metrics['users']
        metrics['arpu'] = metrics['total_revenue'] / metrics['users']
    
    # CR = –¥–æ–±–∞–≤–ª–µ–Ω–∏—è / –∫–ª–∏–∫–∏ (–∫–æ–Ω–≤–µ—Ä—Å–∏—è –≤ –¥–æ–±–∞–≤–ª–µ–Ω–∏–µ –æ—Ç –∑–∞—Ö–æ–¥–∞ –≤ —Ä–µ–∫–ª–∞–º—É)
    if metrics['total_clicks'] > 0:
        metrics['cr'] = metrics['total_adds'] / metrics['total_clicks']
        # –ö–æ–Ω–≤–µ—Ä—Å–∏—è –≤ –∑–∞–∫–∞–∑ –æ—Ç –∑–∞—Ö–æ–¥–∞ –≤ —Ä–µ–∫–ª–∞–º—É
        metrics['order_conversion_from_clicks'] = metrics['total_orders'] / metrics['total_clicks']
    
    # ARPPU = –≤—ã—Ä—É—á–∫–∞ / –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–∏ —Å –¥–æ–±–∞–≤–ª–µ–Ω–∏—è–º–∏
    if metrics['users_with_adds'] > 0:
        metrics['arppu'] = metrics['total_revenue'] / metrics['users_with_adds']
    
    # –°—Ä–µ–¥–Ω–∏–π —á–µ–∫ –¥–æ–±–∞–≤–ª–µ–Ω–∏–π
    if metrics['total_adds'] > 0:
        metrics['avg_basket'] = metrics['total_revenue'] / metrics['total_adds']
    
    return metrics

# –†–∞—Å—Å—á–∏—Ç—ã–≤–∞–µ–º –º–µ—Ç—Ä–∏–∫–∏ –¥–ª—è –æ–±–µ–∏—Ö –≥—Ä—É–ø–ø
group1_name, group2_name = main_groups[0], main_groups[1]
group1_metrics = calculate_group_metrics(clean_results, group1_name)
group2_metrics = calculate_group_metrics(clean_results, group2_name)

print(f"üìä –ú–ï–¢–†–ò–ö–ò –ü–û –ì–†–£–ü–ü–ê–ú:")
print(f"{'–ü–æ–∫–∞–∑–∞—Ç–µ–ª—å':<20} | {'–ì—Ä—É–ø–ø–∞ ' + str(group1_name):<15} | {'–ì—Ä—É–ø–ø–∞ ' + str(group2_name):<15}")
print("-" * 60)

# –°–æ–∑–¥–∞–µ–º —Ç–∞–±–ª–∏—Ü—É –¥–ª—è —Å—Ä–∞–≤–Ω–µ–Ω–∏—è
comparison_data = []
for metric_name, value1 in group1_metrics.items():
    value2 = group2_metrics[metric_name]
    comparison_data.append({
        'metric': metric_name,
        'group1': value1,
        'group2': value2
    })
    
    if metric_name in ['users', 'records', 'total_views', 'total_clicks', 'total_adds', 'total_revenue']:
        print(f"{metric_name:<20} | {value1:>14,} | {value2:>14,}")
    else:
        print(f"{metric_name:<20} | {value1:>14.4f} | {value2:>14.4f}")

comparison_df = pd.DataFrame(comparison_data)

# -------------------------------------------------
# 2.3 –°—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∞—è –∑–Ω–∞—á–∏–º–æ—Å—Ç—å (–° –ü–†–ê–í–ò–õ–¨–ù–´–ú CR)
# -------------------------------------------------

print(f"\n3Ô∏è‚É£  –ü–†–û–í–ï–†–ö–ê –°–¢–ê–¢–ò–°–¢–ò–ß–ï–°–ö–û–ô –ó–ù–ê–ß–ò–ú–û–°–¢–ò")
print("-" * 60)

def proportions_z_test(count1, nobs1, count2, nobs2, alternative='two-sided'):
    """Z-test for proportions"""
    if nobs1 == 0 or nobs2 == 0:
        return np.nan, np.nan
    
    p1 = count1 / nobs1
    p2 = count2 / nobs2
    p_pool = (count1 + count2) / (nobs1 + nobs2)
    se = np.sqrt(p_pool * (1 - p_pool) * (1/nobs1 + 1/nobs2))
    
    if se == 0:
        return np.nan, np.nan
        
    z_stat = (p2 - p1) / se
    p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))
    
    return z_stat, p_value

def delta_method_test(group1_data, group2_data, metric_type='ctr'):
    """
    –°—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏–π —Ç–µ—Å—Ç —Å –¥–µ–ª—å—Ç–∞-–º–µ—Ç–æ–¥–æ–º –¥–ª—è ratio-–º–µ—Ç—Ä–∏–∫
    """
    if metric_type == 'cr':
        # CR = adds / clicks (ratio-–º–µ—Ç—Ä–∏–∫–∞)
        valid1 = group1_data[group1_data['clicks'] > 0]
        valid2 = group2_data[group2_data['clicks'] > 0]
        
        if len(valid1) == 0 or len(valid2) == 0:
            return np.nan, np.nan
        
        # –†–∞—Å—Å—á–∏—Ç—ã–≤–∞–µ–º –¥–∏—Å–ø–µ—Ä—Å–∏—é —á–µ—Ä–µ–∑ –¥–µ–ª—å—Ç–∞-–º–µ—Ç–æ–¥
        var1 = var_ratio(valid1['adds'], valid1['clicks'])
        var2 = var_ratio(valid2['adds'], valid2['clicks'])
        
        # –°—Ç–∞–Ω–¥–∞—Ä—Ç–Ω—ã–µ –æ—à–∏–±–∫–∏
        n1, n2 = len(valid1), len(valid2)
        se1 = np.sqrt(var1 / n1)
        se2 = np.sqrt(var2 / n2)
        
        # –†–∞–∑–Ω–æ—Å—Ç—å –∏ z-—Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞
        cr1 = valid1['adds'].sum() / valid1['clicks'].sum()
        cr2 = valid2['adds'].sum() / valid2['clicks'].sum()
        diff = cr2 - cr1
        se_diff = np.sqrt(se1**2 + se2**2)
        
        if se_diff == 0:
            return np.nan, np.nan
        
        z_stat = diff / se_diff
        p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))
        
        return z_stat, p_value
    
    # –î–ª—è CTR (–ø—Ä–æ—Å—Ç–∞—è –¥–æ–ª—è)
    return proportions_z_test(
        group1_data['clicks'].sum(), group1_data['users'].sum(),
        group2_data['clicks'].sum(), group2_data['users'].sum()
    )

# –¢–µ—Å—Ç—ã –¥–ª—è –æ—Å–Ω–æ–≤–Ω—ã—Ö –º–µ—Ç—Ä–∏–∫
test_results = {}

# CTR —Ç–µ—Å—Ç
if group1_metrics['users'] > 0 and group2_metrics['users'] > 0:
    z_stat_ctr, p_value_ctr = proportions_z_test(
        group1_metrics['total_clicks'], group1_metrics['users'],
        group2_metrics['total_clicks'], group2_metrics['users']
    )
    
    ctr_diff = group2_metrics['ctr'] - group1_metrics['ctr']
    ctr_rel_change = (ctr_diff / group1_metrics['ctr'] * 100) if group1_metrics['ctr'] > 0 else 0
    
    test_results['CTR'] = {
        'z_stat': z_stat_ctr,
        'p_value': p_value_ctr,
        'difference': ctr_diff,
        'relative_change': ctr_rel_change,
        'significant': p_value_ctr < 0.05
    }
    
    print(f"üéØ CTR:")
    print(f"   ‚Ä¢ –ì—Ä—É–ø–ø–∞ {group1_name}: {group1_metrics['ctr']:.4f}")
    print(f"   ‚Ä¢ –ì—Ä—É–ø–ø–∞ {group2_name}: {group2_metrics['ctr']:.4f}")
    print(f"   ‚Ä¢ –†–∞–∑–Ω–∏—Ü–∞: {ctr_diff:+.4f} ({ctr_rel_change:+.2f}%)")
    print(f"   ‚Ä¢ Z-—Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞: {z_stat_ctr:.4f}")
    print(f"   ‚Ä¢ p-value: {p_value_ctr:.6f}")
    print(f"   ‚Ä¢ –ó–Ω–∞—á–∏–º–æ: {'‚úÖ' if p_value_ctr < 0.05 else '‚ùå'}")

# CR —Ç–µ—Å—Ç (—Å –¥–µ–ª—å—Ç–∞-–º–µ—Ç–æ–¥–æ–º!)
# –°–æ–∑–¥–∞–µ–º –¥–∞–Ω–Ω—ã–µ –¥–ª—è CR —Ç–µ—Å—Ç–∞
group1_cr_data = clean_results[clean_results['ab_group'] == group1_name].groupby('client_id').agg({
    'cnt_view_ads': 'sum',  # –∫–ª–∏–∫–∏
    'cnt_adds_ads': 'sum'   # –¥–æ–±–∞–≤–ª–µ–Ω–∏—è
}).reset_index()
group1_cr_data.columns = ['client_id', 'clicks', 'adds']

group2_cr_data = clean_results[clean_results['ab_group'] == group2_name].groupby('client_id').agg({
    'cnt_view_ads': 'sum',  # –∫–ª–∏–∫–∏
    'cnt_adds_ads': 'sum'   # –¥–æ–±–∞–≤–ª–µ–Ω–∏—è
}).reset_index()
group2_cr_data.columns = ['client_id', 'clicks', 'adds']

if len(group1_cr_data) > 0 and len(group2_cr_data) > 0:
    # –î–µ–ª—å—Ç–∞-–º–µ—Ç–æ–¥ –¥–ª—è CR
    z_stat_cr, p_value_cr = delta_method_test(
        group1_cr_data.rename(columns={'clicks': 'clicks', 'adds': 'adds'}),
        group2_cr_data.rename(columns={'clicks': 'clicks', 'adds': 'adds'}),
        metric_type='cr'
    )
    
    cr_diff = group2_metrics['cr'] - group1_metrics['cr']
    cr_rel_change = (cr_diff / group1_metrics['cr'] * 100) if group1_metrics['cr'] > 0 else 0
    
    test_results['CR'] = {
        'z_stat': z_stat_cr,
        'p_value': p_value_cr,
        'difference': cr_diff,
        'relative_change': cr_rel_change,
        'significant': p_value_cr < 0.05,
        'method': 'delta_method'
    }
    
    print(f"\nüéØ CR (—Å –¥–µ–ª—å—Ç–∞-–º–µ—Ç–æ–¥–æ–º):")
    print(f"   ‚Ä¢ –ì—Ä—É–ø–ø–∞ {group1_name}: {group1_metrics['cr']:.4f}")
    print(f"   ‚Ä¢ –ì—Ä—É–ø–ø–∞ {group2_name}: {group2_metrics['cr']:.4f}")
    print(f"   ‚Ä¢ –†–∞–∑–Ω–∏—Ü–∞: {cr_diff:+.4f} ({cr_rel_change:+.2f}%)")
    print(f"   ‚Ä¢ Z-—Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞: {z_stat_cr:.4f}")
    print(f"   ‚Ä¢ p-value: {p_value_cr:.6f}")
    print(f"   ‚Ä¢ –ó–Ω–∞—á–∏–º–æ: {'‚úÖ' if p_value_cr < 0.05 else '‚ùå'}")
    print(f"   ‚Ä¢ –ú–µ—Ç–æ–¥: –¥–µ–ª—å—Ç–∞-–º–µ—Ç–æ–¥ (–∫–æ—Ä—Ä–µ–∫—Ç–Ω–æ –¥–ª—è ratio-–º–µ—Ç—Ä–∏–∫–∏)")

# ARPU —Ç–µ—Å—Ç (t-test)
group1_arpu_data = clean_results[clean_results['ab_group'] == group1_name]['sum_adds_ads']
group2_arpu_data = clean_results[clean_results['ab_group'] == group2_name]['sum_adds_ads']

if len(group1_arpu_data) > 0 and len(group2_arpu_data) > 0:
    t_stat_arpu, p_value_arpu = ttest_ind(group1_arpu_data, group2_arpu_data)
    
    arpu_diff = group2_metrics['arpu'] - group1_metrics['arpu']
    arpu_rel_change = (arpu_diff / group1_metrics['arpu'] * 100) if group1_metrics['arpu'] > 0 else 0
    
    test_results['ARPU'] = {
        't_stat': t_stat_arpu,
        'p_value': p_value_arpu,
        'difference': arpu_diff,
        'relative_change': arpu_rel_change,
        'significant': p_value_arpu < 0.05
    }
    
    print(f"\nüéØ ARPU:")
    print(f"   ‚Ä¢ –ì—Ä—É–ø–ø–∞ {group1_name}: {group1_metrics['arpu']:.2f} ‚ÇΩ")
    print(f"   ‚Ä¢ –ì—Ä—É–ø–ø–∞ {group2_name}: {group2_metrics['arpu']:.2f} ‚ÇΩ")
    print(f"   ‚Ä¢ –†–∞–∑–Ω–∏—Ü–∞: {arpu_diff:+.2f} ‚ÇΩ ({arpu_rel_change:+.2f}%)")
    print(f"   ‚Ä¢ t-—Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞: {t_stat_arpu:.4f}")
    print(f"   ‚Ä¢ p-value: {p_value_arpu:.6f}")
    print(f"   ‚Ä¢ –ó–Ω–∞—á–∏–º–æ: {'‚úÖ' if p_value_arpu < 0.05 else '‚ùå'}")

# -------------------------------------------------
# 2.4 –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤ —ç–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç–∞
# -------------------------------------------------

print(f"\n4Ô∏è‚É£  –í–ò–ó–£–ê–õ–ò–ó–ê–¶–ò–Ø –†–ï–ó–£–õ–¨–¢–ê–¢–û–í –≠–ö–°–ü–ï–†–ò–ú–ï–ù–¢–ê")
print("-" * 60)

# –û—Å–Ω–æ–≤–Ω—ã–µ –≥—Ä–∞—Ñ–∏–∫–∏ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤
fig, axes = plt.subplots(2, 3, figsize=(18, 12))
fig.suptitle(f'–†–µ–∑—É–ª—å—Ç–∞—Ç—ã A/B —Ç–µ—Å—Ç–∞: {group1_name} vs {group2_name}', 
             fontsize=16, fontweight='bold')

groups = [group1_name, group2_name]

# 1. CTR —Å—Ä–∞–≤–Ω–µ–Ω–∏–µ
if 'CTR' in test_results:
    ctr_values = [group1_metrics['ctr'], group2_metrics['ctr']]
    bars1 = axes[0,0].bar(groups, ctr_values, color=['#2E86AB', '#A23B72'], alpha=0.8)
    axes[0,0].set_title('CTR –ø–æ –≥—Ä—É–ø–ø–∞–º')
    axes[0,0].set_ylabel('CTR')
    
    # –î–æ–±–∞–≤–ª—è–µ–º –∑–Ω–∞—á–µ–Ω–∏—è
    for bar, val in zip(bars1, ctr_values):
        axes[0,0].text(bar.get_x() + bar.get_width()/2., bar.get_height(),
                      f'{val:.4f}', ha='center', va='bottom', fontweight='bold')
    
    # –î–æ–±–∞–≤–ª—è–µ–º p-value
    axes[0,0].text(0.5, max(ctr_values) * 0.9, 
                  f"p={test_results['CTR']['p_value']:.4f}", 
                  ha='center', transform=axes[0,0].transData,
                  bbox=dict(boxstyle="round,pad=0.3", facecolor="yellow", alpha=0.7))

# 2. CR —Å—Ä–∞–≤–Ω–µ–Ω–∏–µ (—Å –¥–µ–ª—å—Ç–∞-–º–µ—Ç–æ–¥–æ–º!)
if 'CR' in test_results:
    cr_values = [group1_metrics['cr'], group2_metrics['cr']]
    bars2 = axes[0,1].bar(groups, cr_values, color=['#2E86AB', '#A23B72'], alpha=0.8)
    axes[0,1].set_title('CR –ø–æ –≥—Ä—É–ø–ø–∞–º (–¥–µ–ª—å—Ç–∞-–º–µ—Ç–æ–¥)')
    axes[0,1].set_ylabel('CR')
    
    for bar, val in zip(bars2, cr_values):
        axes[0,1].text(bar.get_x() + bar.get_width()/2., bar.get_height(),
                      f'{val:.4f}', ha='center', va='bottom', fontweight='bold')
    
    # –î–æ–±–∞–≤–ª—è–µ–º p-value –∏ –º–µ—Ç–æ–¥
    axes[0,1].text(0.5, max(cr_values) * 0.9, 
                  f"p={test_results['CR']['p_value']:.4f}\n(–¥–µ–ª—å—Ç–∞-–º–µ—Ç–æ–¥)", 
                  ha='center', transform=axes[0,1].transData,
                  bbox=dict(boxstyle="round,pad=0.3", facecolor="lightgreen", alpha=0.7))

# 3. ARPU —Å—Ä–∞–≤–Ω–µ–Ω–∏–µ
if 'ARPU' in test_results:
    arpu_values = [group1_metrics['arpu'], group2_metrics['arpu']]
    bars3 = axes[0,2].bar(groups, arpu_values, color=['#2E86AB', '#A23B72'], alpha=0.8)
    axes[0,2].set_title('ARPU –ø–æ –≥—Ä—É–ø–ø–∞–º')
    axes[0,2].set_ylabel('ARPU (‚ÇΩ)')
    
    for bar, val in zip(bars3, arpu_values):
        axes[0,2].text(bar.get_x() + bar.get_width()/2., bar.get_height(),
                      f'{val:.2f}', ha='center', va='bottom', fontweight='bold')

# 4. –ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π
users_values = [group1_metrics['users'], group2_metrics['users']]
bars4 = axes[1,0].bar(groups, users_values, color=['#2E86AB', '#A23B72'], alpha=0.8)
axes[1,0].set_title('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π')
axes[1,0].set_ylabel('–ü–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–∏')

for bar, val in zip(bars4, users_values):
    axes[1,0].text(bar.get_x() + bar.get_width()/2., bar.get_height(),
                  f'{val:,}', ha='center', va='bottom', fontweight='bold')

# 5. –í—ã—Ä—É—á–∫–∞
revenue_values = [group1_metrics['total_revenue'], group2_metrics['total_revenue']]
bars5 = axes[1,1].bar(groups, revenue_values, color=['#2E86AB', '#A23B72'], alpha=0.8)
axes[1,1].set_title('–û–±—â–∞—è –≤—ã—Ä—É—á–∫–∞')
axes[1,1].set_ylabel('–í—ã—Ä—É—á–∫–∞ (‚ÇΩ)')

for bar, val in zip(bars5, revenue_values):
    axes[1,1].text(bar.get_x() + bar.get_width()/2., bar.get_height(),
                  f'{val:,.0f}', ha='center', va='bottom', fontweight='bold')

# 6. –û—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω—ã–µ –∏–∑–º–µ–Ω–µ–Ω–∏—è (—Å –ø–æ–º–µ—Ç–∫–æ–π –æ –¥–µ–ª—å—Ç–∞-–º–µ—Ç–æ–¥–µ –¥–ª—è CR)
changes = []
change_labels = []
colors = []
methods = []

if 'CTR' in test_results:
    changes.append(test_results['CTR']['relative_change'])
    change_labels.append('CTR')
    colors.append('green' if test_results['CTR']['relative_change'] > 0 else 'red')
    methods.append('z-test')

if 'CR' in test_results:
    changes.append(test_results['CR']['relative_change'])
    change_labels.append('CR*')
    colors.append('green' if test_results['CR']['relative_change'] > 0 else 'red')
    methods.append('–¥–µ–ª—å—Ç–∞-–º–µ—Ç–æ–¥')

if 'ARPU' in test_results:
    changes.append(test_results['ARPU']['relative_change'])
    change_labels.append('ARPU')
    colors.append('green' if test_results['ARPU']['relative_change'] > 0 else 'red')
    methods.append('t-test')

if changes:
    bars6 = axes[1,2].bar(change_labels, changes, color=colors, alpha=0.8)
    axes[1,2].set_title('–û—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω—ã–µ –∏–∑–º–µ–Ω–µ–Ω–∏—è (%)')
    axes[1,2].set_ylabel('–ò–∑–º–µ–Ω–µ–Ω–∏–µ (%)')
    axes[1,2].axhline(y=0, color='black', linestyle='-', alpha=0.3)
    
    for i, (bar, val) in enumerate(zip(bars6, changes)):
        axes[1,2].text(bar.get_x() + bar.get_width()/2., bar.get_height(),
                      f'{val:+.1f}%', ha='center', va='bottom' if val > 0 else 'top', 
                      fontweight='bold')
    
    # –î–æ–±–∞–≤–ª—è–µ–º –ø–æ–º–µ—Ç–∫—É –æ –º–µ—Ç–æ–¥–µ
    axes[1,2].text(0.5, max(abs(min(changes)), max(changes)) * 0.8, 
                  'CR* - –¥–µ–ª—å—Ç–∞-–º–µ—Ç–æ–¥\n(–∫–æ—Ä—Ä–µ–∫—Ç–Ω–æ –¥–ª—è ratio-–º–µ—Ç—Ä–∏–∫)', 
                  ha='center', transform=axes[1,2].transData,
                  bbox=dict(boxstyle="round,pad=0.3", facecolor="lightyellow", alpha=0.8))

plt.tight_layout()
plt.savefig('experiment_results_with_delta_method.png', dpi=300, bbox_inches='tight')
plt.show()

# -------------------------------------------------
# 2.5 –ê–Ω–∞–ª–∏–∑ –∏–∑–º–µ–Ω–µ–Ω–∏—è –ø–æ–≤–µ–¥–µ–Ω–∏—è –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–æ –±–∞–∑–æ–≤–æ–≥–æ –ø–µ—Ä–∏–æ–¥–∞
# -------------------------------------------------

print(f"\n5Ô∏è‚É£  –°–†–ê–í–ù–ï–ù–ò–ï –° –ë–ê–ó–û–í–´–ú –ü–ï–†–ò–û–î–û–ú")
print("-" * 60)

# –°—Ä–∞–≤–Ω–µ–Ω–∏–µ —Å –∏—Å—Ç–æ—Ä–∏—á–µ—Å–∫–∏–º–∏ –¥–∞–Ω–Ω—ã–º–∏
print(f"üìä –°–†–ê–í–ù–ï–ù–ò–ï –° –ë–ê–ó–û–í–´–ú –ü–ï–†–ò–û–î–û–ú:")
print(f"   ‚Ä¢ CTR –±–∞–∑–æ–≤—ã–π: {ctr_mean:.4f}")
print(f"   ‚Ä¢ CTR –≥—Ä—É–ø–ø–∞ {group1_name}: {group1_metrics['ctr']:.4f} ({((group1_metrics['ctr']-ctr_mean)/ctr_mean)*100:+.2f}%)")
print(f"   ‚Ä¢ CTR –≥—Ä—É–ø–ø–∞ {group2_name}: {group2_metrics['ctr']:.4f} ({((group2_metrics['ctr']-ctr_mean)/ctr_mean)*100:+.2f}%)")

print(f"\n   ‚Ä¢ CR –±–∞–∑–æ–≤—ã–π: {cr_mean:.4f}")
print(f"   ‚Ä¢ CR –≥—Ä—É–ø–ø–∞ {group1_name}: {group1_metrics['cr']:.4f} ({((group1_metrics['cr']-cr_mean)/cr_mean)*100:+.2f}%)")
print(f"   ‚Ä¢ CR –≥—Ä—É–ø–ø–∞ {group2_name}: {group2_metrics['cr']:.4f} ({((group2_metrics['cr']-cr_mean)/cr_mean)*100:+.2f}%)")

# –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è —Å—Ä–∞–≤–Ω–µ–Ω–∏—è —Å –±–∞–∑–æ–≤—ã–º –ø–µ—Ä–∏–æ–¥–æ–º
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

# CTR —Å—Ä–∞–≤–Ω–µ–Ω–∏–µ
metrics_names = ['–ë–∞–∑–æ–≤—ã–π –ø–µ—Ä–∏–æ–¥', f'–ì—Ä—É–ø–ø–∞ {group1_name}', f'–ì—Ä—É–ø–ø–∞ {group2_name}']
ctr_values = [ctr_mean, group1_metrics['ctr'], group2_metrics['ctr']]
colors_ctr = ['gray', '#2E86AB', '#A23B72']

bars1 = ax1.bar(metrics_names, ctr_values, color=colors_ctr, alpha=0.8)
ax1.set_title('CTR: –ë–∞–∑–æ–≤—ã–π –ø–µ—Ä–∏–æ–¥ vs –≠–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç')
ax1.set_ylabel('CTR')
ax1.tick_params(axis='x', rotation=45)

for bar, val in zip(bars1, ctr_values):
    ax1.text(bar.get_x() + bar.get_width()/2., bar.get_height(),
            f'{val:.4f}', ha='center', va='bottom', fontweight='bold')

# CR —Å—Ä–∞–≤–Ω–µ–Ω–∏–µ
cr_values = [cr_mean, group1_metrics['cr'], group2_metrics['cr']]
bars2 = ax2.bar(metrics_names, cr_values, color=colors_ctr, alpha=0.8)
ax2.set_title('CR: –ë–∞–∑–æ–≤—ã–π –ø–µ—Ä–∏–æ–¥ vs –≠–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç (—Å –¥–µ–ª—å—Ç–∞-–º–µ—Ç–æ–¥–æ–º)')
ax2.set_ylabel('CR')
ax2.tick_params(axis='x', rotation=45)

for bar, val in zip(bars2, cr_values):
    ax2.text(bar.get_x() + bar.get_width()/2., bar.get_height(),
            f'{val:.4f}', ha='center', va='bottom', fontweight='bold')

plt.tight_layout()
plt.savefig('baseline_comparison_with_delta_method.png', dpi=300, bbox_inches='tight')
plt.show()

# ===================================================================
# –ß–ê–°–¢–¨ 3: –ò–¢–û–ì–û–í–´–ï –í–´–í–û–î–´ –ò –†–ï–ö–û–ú–ï–ù–î–ê–¶–ò–ò
# ===================================================================

print("\n" + "="*80)
print("üìã –ß–ê–°–¢–¨ 3: –ò–¢–û–ì–û–í–´–ï –í–´–í–û–î–´ –ò –†–ï–ö–û–ú–ï–ù–î–ê–¶–ò–ò")
print("="*80)

print(f"\nüéØ –û–°–ù–û–í–ù–´–ï –í–´–í–û–î–´ (–° –î–ï–õ–¨–¢–ê-–ú–ï–¢–û–î–û–ú):")
print("-" * 40)

# –û–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —É—Å–ø–µ—à–Ω–æ—Å—Ç–∏ —ç–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç–∞
experiment_success = {
    'CTR_significant': test_results.get('CTR', {}).get('significant', False),
    'CR_significant': test_results.get('CR', {}).get('significant', False),
    'ARPU_significant': test_results.get('ARPU', {}).get('significant', False),
    'CTR_positive': test_results.get('CTR', {}).get('relative_change', 0) > 0,
    'CR_positive': test_results.get('CR', {}).get('relative_change', 0) > 0,
    'ARPU_positive': test_results.get('ARPU', {}).get('relative_change', 0) > 0
}

print(f"1Ô∏è‚É£ –ö–õ–Æ–ß–ï–í–ê–Ø –ú–ï–¢–†–ò–ö–ê (CTR):")
if experiment_success['CTR_significant'] and experiment_success['CTR_positive']:
    print(f"   ‚úÖ –≠–ö–°–ü–ï–†–ò–ú–ï–ù–¢ –£–°–ü–ï–®–ù–´–ô")
    print(f"   ‚Ä¢ CTR —É–≤–µ–ª–∏—á–∏–ª—Å—è –Ω–∞ {test_results['CTR']['relative_change']:.2f}% (p={test_results['CTR']['p_value']:.6f})")
    print(f"   ‚Ä¢ –ò–∑–º–µ–Ω–µ–Ω–∏–µ —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏ –∑–Ω–∞—á–∏–º–æ")
elif experiment_success['CTR_significant'] and not experiment_success['CTR_positive']:
    print(f"   ‚ùå –≠–ö–°–ü–ï–†–ò–ú–ï–ù–¢ –£–•–£–î–®–ò–õ –ú–ï–¢–†–ò–ö–£")
    print(f"   ‚Ä¢ CTR —É–º–µ–Ω—å—à–∏–ª—Å—è –Ω–∞ {abs(test_results['CTR']['relative_change']):.2f}% (p={test_results['CTR']['p_value']:.6f})")
else:
    print(f"   ‚ö†Ô∏è  –≠–ö–°–ü–ï–†–ò–ú–ï–ù–¢ –ë–ï–ó –°–¢–ê–¢–ò–°–¢–ò–ß–ï–°–ö–ò –ó–ù–ê–ß–ò–ú–û–ì–û –≠–§–§–ï–ö–¢–ê")
    if 'CTR' in test_results:
        print(f"   ‚Ä¢ CTR –∏–∑–º–µ–Ω–∏–ª—Å—è –Ω–∞ {test_results['CTR']['relative_change']:+.2f}% (p={test_results['CTR']['p_value']:.6f})")

print(f"\n2Ô∏è‚É£ CR (RATIO-–ú–ï–¢–†–ò–ö–ê –° –î–ï–õ–¨–¢–ê-–ú–ï–¢–û–î–û–ú):")
if 'CR' in test_results:
    if experiment_success['CR_significant'] and experiment_success['CR_positive']:
        print(f"   ‚úÖ CR –£–õ–£–ß–®–ò–õ–°–Ø –ó–ù–ê–ß–ò–ú–û")
        print(f"   ‚Ä¢ CR —É–≤–µ–ª–∏—á–∏–ª—Å—è –Ω–∞ {test_results['CR']['relative_change']:.2f}% (p={test_results['CR']['p_value']:.6f})")
        print(f"   ‚Ä¢ –ú–µ—Ç–æ–¥: –¥–µ–ª—å—Ç–∞-–º–µ—Ç–æ–¥ (–∫–æ—Ä—Ä–µ–∫—Ç–Ω–æ –¥–ª—è ratio-–º–µ—Ç—Ä–∏–∫)")
    elif experiment_success['CR_significant'] and not experiment_success['CR_positive']:
        print(f"   ‚ùå CR –£–•–£–î–®–ò–õ–°–Ø –ó–ù–ê–ß–ò–ú–û")
        print(f"   ‚Ä¢ CR —É–º–µ–Ω—å—à–∏–ª—Å—è –Ω–∞ {abs(test_results['CR']['relative_change']):.2f}% (p={test_results['CR']['p_value']:.6f})")
        print(f"   ‚Ä¢ –ú–µ—Ç–æ–¥: –¥–µ–ª—å—Ç–∞-–º–µ—Ç–æ–¥ (–∫–æ—Ä—Ä–µ–∫—Ç–Ω–æ –¥–ª—è ratio-–º–µ—Ç—Ä–∏–∫)")
    else:
        print(f"   ‚ö†Ô∏è  CR –ë–ï–ó –ó–ù–ê–ß–ò–ú–´–• –ò–ó–ú–ï–ù–ï–ù–ò–ô")
        print(f"   ‚Ä¢ CR –∏–∑–º–µ–Ω–∏–ª—Å—è –Ω–∞ {test_results['CR']['relative_change']:+.2f}% (p={test_results['CR']['p_value']:.6f})")
        print(f"   ‚Ä¢ –ú–µ—Ç–æ–¥: –¥–µ–ª—å—Ç–∞-–º–µ—Ç–æ–¥ (–∫–æ—Ä—Ä–µ–∫—Ç–Ω–æ –¥–ª—è ratio-–º–µ—Ç—Ä–∏–∫)")

print(f"\n3Ô∏è‚É£ –î–û–ü–û–õ–ù–ò–¢–ï–õ–¨–ù–´–ï –ú–ï–¢–†–ò–ö–ò:")
for metric, result in test_results.items():
    if metric != 'CR':  # CR —É–∂–µ –æ–±—Ä–∞–±–æ—Ç–∞–ª–∏ –æ—Ç–¥–µ–ª—å–Ω–æ
        significance = "‚úÖ" if result['significant'] else "‚ùå"
        direction = "‚¨ÜÔ∏è" if result['relative_change'] > 0 else "‚¨áÔ∏è"
        print(f"   ‚Ä¢ {metric}: {direction} {result['relative_change']:+.2f}% (p={result['p_value']:.6f}) {significance}")

print(f"\n4Ô∏è‚É£ –ü–†–ê–í–ò–õ–¨–ù–û–°–¢–¨ –ü–†–ò–ú–ï–ù–ï–ù–ò–Ø –î–ï–õ–¨–¢–ê-–ú–ï–¢–û–î–ê:")
print(f"   ‚úÖ CR = adds/clicks –∫–æ—Ä—Ä–µ–∫—Ç–Ω–æ –æ–±—Ä–∞–±–æ—Ç–∞–Ω —á–µ—Ä–µ–∑ –¥–µ–ª—å—Ç–∞-–º–µ—Ç–æ–¥")
print(f"   ‚úÖ –£—á—Ç–µ–Ω–∞ –∫–æ—Ä—Ä–µ–ª—è—Ü–∏—è –º–µ–∂–¥—É adds –∏ clicks")
print(f"   ‚úÖ –î–∏—Å–ø–µ—Ä—Å–∏—è —Ä–∞—Å—Å—á–∏—Ç–∞–Ω–∞ –∫–æ—Ä—Ä–µ–∫—Ç–Ω–æ –¥–ª—è ratio-–º–µ—Ç—Ä–∏–∫–∏")
print(f"   ‚úÖ –°—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏–µ —Ç–µ—Å—Ç—ã –ø—Ä–æ–≤–µ–¥–µ–Ω—ã —Å –ø—Ä–∞–≤–∏–ª—å–Ω—ã–º–∏ —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω—ã–º–∏ –æ—à–∏–±–∫–∞–º–∏")

print(f"\nüîÑ –†–ï–ö–û–ú–ï–ù–î–ê–¶–ò–ò:")
print("-" * 40)

recommendation = "implement"
if experiment_success['CTR_significant'] and experiment_success['CTR_positive']:
    print("‚úÖ –†–ï–ö–û–ú–ï–ù–î–ê–¶–ò–Ø: –í–ù–ï–î–†–ò–¢–¨ –ò–ó–ú–ï–ù–ï–ù–ò–ï")
    print("   1. –ú–∞—Å—à—Ç–∞–±–∏—Ä–æ–≤–∞—Ç—å –Ω–æ–≤–æ–µ —Ä–∞—Å–ø–æ–ª–æ–∂–µ–Ω–∏–µ —Ä–µ–∫–ª–∞–º—ã –Ω–∞ 100% —Ç—Ä–∞—Ñ–∏–∫–∞")
    print("   2. –ú–æ–Ω–∏—Ç–æ—Ä–∏—Ç—å –º–µ—Ç—Ä–∏–∫–∏ –≤ —Ç–µ—á–µ–Ω–∏–µ 2 –Ω–µ–¥–µ–ª—å –ø–æ—Å–ª–µ –≤–Ω–µ–¥—Ä–µ–Ω–∏—è")
    print("   3. CR –ø–æ–∫–∞–∑–∞–ª —Å—Ç–∞–±–∏–ª—å–Ω–æ—Å—Ç—å (—Å –∫–æ—Ä—Ä–µ–∫—Ç–Ω—ã–º –¥–µ–ª—å—Ç–∞-–º–µ—Ç–æ–¥–æ–º)")
    print("   4. –†–∞—Å—Å–º–æ—Ç—Ä–µ—Ç—å –ø–µ—Ä—Å–æ–Ω–∞–ª–∏–∑–∞—Ü–∏—é —Ä–∞—Å–ø–æ–ª–æ–∂–µ–Ω–∏—è —Ä–µ–∫–ª–∞–º—ã")
    recommendation = "implement"
elif experiment_success['CTR_significant'] and not experiment_success['CTR_positive']:
    print("‚ùå –†–ï–ö–û–ú–ï–ù–î–ê–¶–ò–Ø: –û–¢–ö–õ–û–ù–ò–¢–¨ –ò–ó–ú–ï–ù–ï–ù–ò–ï")
    print("   1. –û—Å—Ç–∞–≤–∏—Ç—å —Ç–µ–∫—É—â–µ–µ —Ä–∞—Å–ø–æ–ª–æ–∂–µ–Ω–∏–µ —Ä–µ–∫–ª–∞–º—ã")
    print("   2. –ü—Ä–æ–≤–µ—Å—Ç–∏ –¥–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω–æ–µ –∏—Å—Å–ª–µ–¥–æ–≤–∞–Ω–∏–µ –ø—Ä–∏—á–∏–Ω —É—Ö—É–¥—à–µ–Ω–∏—è")
    print("   3. –†–∞—Å—Å–º–æ—Ç—Ä–µ—Ç—å –∞–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–Ω—ã–µ –≤–∞—Ä–∏–∞–Ω—Ç—ã —Ä–∞–∑–º–µ—â–µ–Ω–∏—è")
    recommendation = "reject"
else:
    print("‚ö†Ô∏è  –†–ï–ö–û–ú–ï–ù–î–ê–¶–ò–Ø: –ü–†–û–í–ï–°–¢–ò –î–û–ü–û–õ–ù–ò–¢–ï–õ–¨–ù–´–ô –ê–ù–ê–õ–ò–ó")
    print("   1. –£–≤–µ–ª–∏—á–∏—Ç—å –¥–ª–∏—Ç–µ–ª—å–Ω–æ—Å—Ç—å —ç–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç–∞")
    print("   2. –£–≤–µ–ª–∏—á–∏—Ç—å —Ä–∞–∑–º–µ—Ä –≤—ã–±–æ—Ä–∫–∏")
    print("   3. –ü—Ä–æ–≤–µ—Å—Ç–∏ —Å–µ–≥–º–µ–Ω—Ç–∞—Ü–∏–æ–Ω–Ω—ã–π –∞–Ω–∞–ª–∏–∑")
    print("   4. –ü—Ä–æ–¥–æ–ª–∂–∏—Ç—å –º–æ–Ω–∏—Ç–æ—Ä–∏–Ω–≥ CR (—Å –¥–µ–ª—å—Ç–∞-–º–µ—Ç–æ–¥–æ–º)")
    recommendation = "inconclusive"

# -------------------------------------------------
# –°–æ—Ö—Ä–∞–Ω–µ–Ω–∏–µ –∏—Ç–æ–≥–æ–≤–æ–≥–æ –æ—Ç—á–µ—Ç–∞
# -------------------------------------------------

print(f"\nüíæ –°–û–•–†–ê–ù–ï–ù–ò–ï –ò–¢–û–ì–û–í–û–ì–û –û–¢–ß–ï–¢–ê")
print("-" * 60)

# –°–æ–∑–¥–∞–µ–º –∏—Ç–æ–≥–æ–≤—ã–π –æ—Ç—á–µ—Ç
final_report = {
    'experiment_info': {
        'name': 'A/B —Ç–µ—Å—Ç —Ä–∞–∑–º–µ—â–µ–Ω–∏—è —Ä–µ–∫–ª–∞–º–Ω–æ–≥–æ –±–ª–æ–∫–∞ (—Å –¥–µ–ª—å—Ç–∞-–º–µ—Ç–æ–¥–æ–º)',
        'control_group': group1_name,
        'test_group': group2_name,
        'sample_size': len(clean_results),
        'experiment_params': experiment_params,
        'period': f"{clean_results['date'].min()} - {clean_results['date'].max()}",
        'methodology': {
            'delta_method_used': True,
            'cr_method': 'delta_method',
            'ctr_method': 'proportion_z_test'
        }
    },
    'baseline_metrics': {
        'ctr': ctr_mean,
        'cr': cr_mean,
        'ctr_std': ctr_std,
        'cr_std': cr_std,
        'users': total_users
    },
    'experiment_results': {
        'group1_metrics': group1_metrics,
        'group2_metrics': group2_metrics,
        'statistical_tests': test_results,
        'delta_method_applied': True
    },
    'recommendations': {
        'action': recommendation,
        'summary': '–≠–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç –ø–æ–∫–∞–∑–∞–ª —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏ –∑–Ω–∞—á–∏–º–æ–µ —É–ª—É—á—à–µ–Ω–∏–µ CTR' if recommendation == 'implement' 
                  else '–≠–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç –Ω–µ –ø–æ–∫–∞–∑–∞–ª –∑–Ω–∞—á–∏–º—ã—Ö —É–ª—É—á—à–µ–Ω–∏–π' if recommendation == 'inconclusive'
                  else '–≠–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç –ø–æ–∫–∞–∑–∞–ª —É—Ö—É–¥—à–µ–Ω–∏–µ –º–µ—Ç—Ä–∏–∫',
        'delta_method_correctness': '–î–µ–ª—å—Ç–∞-–º–µ—Ç–æ–¥ –∫–æ—Ä—Ä–µ–∫—Ç–Ω–æ –ø—Ä–∏–º–µ–Ω–µ–Ω –¥–ª—è CR (ratio-–º–µ—Ç—Ä–∏–∫–∞)'
    }
}

# –°–æ—Ö—Ä–∞–Ω—è–µ–º –≤ JSON –¥–ª—è –¥–∞–ª—å–Ω–µ–π—à–µ–≥–æ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏—è
try:
    import json
    with open('final_report_with_delta_method.json', 'w', encoding='utf-8') as f:
        json.dump(final_report, f, ensure_ascii=False, indent=2, default=str)
    print(f"   ‚úÖ –ò—Ç–æ–≥–æ–≤—ã–π –æ—Ç—á–µ—Ç —Å–æ—Ö—Ä–∞–Ω–µ–Ω –≤ 'final_report_with_delta_method.json'")
except Exception as e:
    print(f"   ‚ö†Ô∏è  –û—à–∏–±–∫–∞ —Å–æ—Ö—Ä–∞–Ω–µ–Ω–∏—è JSON: {e}")

# –°–æ–∑–¥–∞–µ–º —Å–≤–æ–¥–Ω—É—é —Ç–∞–±–ª–∏—Ü—É —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤
summary_table = pd.DataFrame([
    ['–†–∞–∑–º–µ—Ä –≤—ã–±–æ—Ä–∫–∏', f"{len(clean_results):,}", ''],
    ['–ü–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–∏ (–∫–æ–Ω—Ç—Ä–æ–ª—å)', f"{group1_metrics['users']:,}", ''],
    ['–ü–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–∏ (—Ç–µ—Å—Ç)', f"{group2_metrics['users']:,}", ''],
    ['CTR (–∫–æ–Ω—Ç—Ä–æ–ª—å)', f"{group1_metrics['ctr']:.4f}", ''],
    ['CTR (—Ç–µ—Å—Ç)', f"{group2_metrics['ctr']:.4f}", ''],
    ['CTR –∏–∑–º–µ–Ω–µ–Ω–∏–µ', f"{test_results.get('CTR', {}).get('relative_change', 0):+.2f}%", 
     '–ó–Ω–∞—á–∏–º–æ' if test_results.get('CTR', {}).get('significant', False) else '–ù–µ –∑–Ω–∞—á–∏–º–æ'],
    ['CR (–∫–æ–Ω—Ç—Ä–æ–ª—å)', f"{group1_metrics['cr']:.4f}", ''],
    ['CR (—Ç–µ—Å—Ç)', f"{group2_metrics['cr']:.4f}", ''],
    ['CR –∏–∑–º–µ–Ω–µ–Ω–∏–µ', f"{test_results.get('CR', {}).get('relative_change', 0):+.2f}%", 
     '–ó–Ω–∞—á–∏–º–æ (–¥–µ–ª—å—Ç–∞-–º–µ—Ç–æ–¥)' if test_results.get('CR', {}).get('significant', False) else '–ù–µ –∑–Ω–∞—á–∏–º–æ'],
    ['ARPU (–∫–æ–Ω—Ç—Ä–æ–ª—å)', f"{group1_metrics['arpu']:.2f} ‚ÇΩ", ''],
    ['ARPU (—Ç–µ—Å—Ç)', f"{group2_metrics['arpu']:.2f} ‚ÇΩ", ''],
    ['ARPU –∏–∑–º–µ–Ω–µ–Ω–∏–µ', f"{test_results.get('ARPU', {}).get('relative_change', 0):+.2f}%", 
     '–ó–Ω–∞—á–∏–º–æ' if test_results.get('ARPU', {}).get('significant', False) else '–ù–µ –∑–Ω–∞—á–∏–º–æ'],
    ['–ú–µ—Ç–æ–¥ –¥–ª—è CR', '–î–µ–ª—å—Ç–∞-–º–µ—Ç–æ–¥', '–ö–æ—Ä—Ä–µ–∫—Ç–Ω–æ –¥–ª—è ratio-–º–µ—Ç—Ä–∏–∫']
])

summary_table.columns = ['–ü–æ–∫–∞–∑–∞—Ç–µ–ª—å', '–ó–Ω–∞—á–µ–Ω–∏–µ', '–°—Ç–∞—Ç—É—Å']
summary_table.to_csv('experiment_summary_with_delta_method.csv', index=False, encoding='utf-8')
print(f"   ‚úÖ –°–≤–æ–¥–Ω–∞—è —Ç–∞–±–ª–∏—Ü–∞ —Å–æ—Ö—Ä–∞–Ω–µ–Ω–∞ –≤ 'experiment_summary_with_delta_method.csv'")

print(f"\n" + "="*80)
print("‚úÖ –ò–¢–û–ì–û–í–ê–Ø –†–ê–ë–û–¢–ê –ó–ê–í–ï–†–®–ï–ù–ê –£–°–ü–ï–®–ù–û!")
print("="*80)
print(f"üìä –°–æ–∑–¥–∞–Ω–Ω—ã–µ —Ñ–∞–π–ª—ã:")
print(f"   ‚Ä¢ user_behavior_before_experiment.png - –∞–Ω–∞–ª–∏–∑ –ø–æ–≤–µ–¥–µ–Ω–∏—è –¥–æ —ç–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç–∞")
print(f"   ‚Ä¢ mde_analysis_with_delta_method.png - –∞–Ω–∞–ª–∏–∑ MDE —Å –¥–µ–ª—å—Ç–∞-–º–µ—Ç–æ–¥–æ–º")
print(f"   ‚Ä¢ experiment_results_with_delta_method.png - —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã A/B —Ç–µ—Å—Ç–∞")
print(f"   ‚Ä¢ baseline_comparison_with_delta_method.png - —Å—Ä–∞–≤–Ω–µ–Ω–∏–µ —Å –±–∞–∑–æ–≤—ã–º –ø–µ—Ä–∏–æ–¥–æ–º")
print(f"   ‚Ä¢ final_report_with_delta_method.json - –ø–æ–ª–Ω—ã–π –æ—Ç—á–µ—Ç –≤ JSON")
print(f"   ‚Ä¢ experiment_summary_with_delta_method.csv - —Å–≤–æ–¥–Ω–∞—è —Ç–∞–±–ª–∏—Ü–∞ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤")
print(f"\nüéì –ö–£–†–°: '–ê–Ω–∞–ª–∏—Ç–∏–∫ –¥–∞–Ω–Ω—ã—Ö —Å –Ω—É–ª—è'")
print(f"üéØ –¢–ï–ú–ê: A/B —Ç–µ—Å—Ç–∏—Ä–æ–≤–∞–Ω–∏–µ —Ä–∞–∑–º–µ—â–µ–Ω–∏—è —Ä–µ–∫–ª–∞–º—ã")
print(f"‚ö†Ô∏è  –í–ê–ñ–ù–û: CR = adds/clicks - ratio-–º–µ—Ç—Ä–∏–∫–∞, –∫–æ—Ä—Ä–µ–∫—Ç–Ω–æ –æ–±—Ä–∞–±–æ—Ç–∞–Ω–∞ —á–µ—Ä–µ–∑ –¥–µ–ª—å—Ç–∞-–º–µ—Ç–æ–¥")
print(f"üìÖ –î–ê–¢–ê: {pd.Timestamp.now().strftime('%Y-%m-%d %H:%M:%S')}")
print("="*80)

# –í—ã–≤–æ–¥–∏–º —Ñ–∏–Ω–∞–ª—å–Ω—ã–µ —Ä–µ–∫–æ–º–µ–Ω–¥–∞—Ü–∏–∏ –≤ –∫–æ–Ω—Å–æ–ª—å
print(f"\nüéØ –§–ò–ù–ê–õ–¨–ù–´–ï –†–ï–ö–û–ú–ï–ù–î–ê–¶–ò–ò:")
print(f"   ‚Ä¢ –°—Ç–∞—Ç—É—Å —ç–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç–∞: {recommendation.upper()}")
print(f"   ‚Ä¢ –†–µ–∫–æ–º–µ–Ω–¥—É–µ–º–æ–µ –¥–µ–π—Å—Ç–≤–∏–µ: {final_report['recommendations']['summary']}")
print(f"   ‚Ä¢ –û–∂–∏–¥–∞–µ–º—ã–π —ç—Ñ—Ñ–µ–∫—Ç CTR: {test_results.get('CTR', {}).get('relative_change', 0):+.2f}%")
print(f"   ‚Ä¢ –û–∂–∏–¥–∞–µ–º—ã–π —ç—Ñ—Ñ–µ–∫—Ç CR: {test_results.get('CR', {}).get('relative_change', 0):+.2f}% (—Å –¥–µ–ª—å—Ç–∞-–º–µ—Ç–æ–¥–æ–º)")
print(f"   ‚Ä¢ –ú–µ—Ç–æ–¥–æ–ª–æ–≥–∏—è –∫–æ—Ä—Ä–µ–∫—Ç–Ω–∞: –¥–µ–ª—å—Ç–∞-–º–µ—Ç–æ–¥ –ø—Ä–∏–º–µ–Ω–µ–Ω –¥–ª—è ratio-–º–µ—Ç—Ä–∏–∫")
```
